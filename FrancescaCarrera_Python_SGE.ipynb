### Introduzione.

#### Focus del progetto.

Questo progetto è stato sviluppato per un cliente ipotetico ma realistico: **Voci**.  

Voci è un media brand nato con l’obiettivo di riequilibrare la narrazione mediatica,  
ancora oggi spesso dominata da prospettive maschili.  
La missione di Voci è quella di dare spazio e visibilità alle donne  
che, con piccoli e grandi passi, stanno contribuendo a cambiare il mondo.

Nel loro lavoro editoriale, che include blog, social media e YouTube,  
le fondatrici sentivano il bisogno di uno strumento strutturato  
per organizzare e analizzare i contenuti pubblicati,  
in modo da valorizzare al meglio le diverse forme di narrazione:  
testimonianze, approfondimenti, atti di attivismo.

In risposta a questa esigenza, ho sviluppato un **Sistema di Gestione degli Elementi (SGE)**.  
Il sistema consente agli utenti di:

- Effettuare ricerche selezionando uno o più criteri;
- Visualizzare l’intero dataset, scegliendo l’elemento da usare come ordinamento principale;
- Consultare insight e analisi utili su aspetti statistici;
- Registrare nuovi contenuti, organizzandoli tematicamente.

Ho creato un dizionario di liste di dizionari, rinominato *gender_media_dataset_voci*.  
Le chiavi del dizionario rappresentano le **categorie** tematiche individuate.  
Ogni valore è una lista di articoli, organizzati come sottodizionari contenenti:  
**Posizionamento**, **TipoNarrativa** e **Titolo**.  
La struttura consente una consultazione ordinata dei contenuti e facilita operazioni di ricerca e analisi.  

Questo progetto mi ha permesso di mettere in pratica  
l’utilizzo di strutture dati complesse, funzioni, cicli, logiche condizionali,  
e la progettazione di un’applicazione interamente modulare.  
Ho realizzato diverse classi con responsabilità distinte (controller, modelli, viste),  
curando la gestione del flusso utente, la validazione degli input,  
e l’organizzazione dei contenuti per favorirne la consultazione, la registrazione e l’analisi.  
L’intero sistema è sviluppato in Python puro, senza librerie esterne,  
ad eccezione di importazioni standard minime ove strettamente necessarie.  

#### Gestione degli errori.

In questo progetto non vengono sollevate eccezioni legate alla struttura del dataset,  
perché ho scelto di partire dal presupposto che *gender_media_dataset_voci* sia già corretto e valido.  

Non ho quindi inserito controlli interni per verificare l'esattezza di chiavi e valori.  
La registrazione di nuovi articoli avviene però in modo controllato:  
- Alcuni campi (come il posizionamento) vengono assegnati automaticamente;  
- Altri (come tipo di narrativa e categoria) sono selezionabili da un elenco predefinito;  
- I campi più liberi (es. il titolo) seguono regole base su punteggiatura, stile e formattazione.  

Anche se non posso garantire la correttezza grammaticale, questi accorgimenti aiutano a mantenere il dataset ordinato.  

Ho invece gestito con cura gli input dell’utente all’interno dei menu e delle funzioni principali:  
in caso di inserimenti errati o non previsti, il sistema avvisa l’utente e richiede un nuovo input,  
evitando che il programma si blocchi.  

#### Architettura del progetto.

Il progetto è strutturato secondo un principio ispirato al pattern MVC (Model-View-Controller),  
pur senza implementarlo in modo rigido. Le classi sono suddivise in tre gruppi principali:  

- **Model**: classi responsabili della logica dei dati, delle operazioni sul dataset e delle funzionalità tematiche  
(es. ricerca, registrazione, analisi statistiche).  
- **View**: classi che gestiscono l'interazione testuale con l’utente e la visualizzazione dei risultati.  
- **Controller**: classi che coordinano il flusso logico del programma,  
ricevendo l’input dell’utente e invocando le operazioni corrispondenti nei model.  

Questa suddivisione ha l'obiettivo di migliorare la chiarezza, la riusabilità del codice e la separazione delle responsabilità.  

Tutte le classi (Model, View, Controller) ricevono in input, al momento dell’istanza,  
il parametro *dataset_articoli*, che rappresenta il dataset su cui operare.  
Ho adottato questo approccio per mantenere il progetto flessibile:  
sebbene attualmente si utilizzi un dataset specifico definito internamente,  
è possibile sostituirlo con un altro dataset, purché rispetti la struttura e i requisiti del dataset originale.  

La classe *HomeController* funge da **facade** del progetto:  
rappresenta il punto d'ingresso unico per l'utente, inoltre fa parte del gruppo di classi Controller  
che gestiscono l'interazione tra il gruppo di classi View e Model.      
Tutti i metodi e attributi delle altre classi sono privati per mantenere l'incapsulamento  e permettere  
all'utente di interagire con il sistema esclusivamente tramite il metodo pubblico *avvia()* di *HomeController*.

### Classi Model

#### Introduzione.

In questa sezione sono raccolte le classi che costituiscono il cuore logico del sistema:  
gestiscono i dati e le operazioni tematiche che possono essere eseguite sul dataset degli articoli.  
  
Tutte le classi Model condividono una struttura comune che consente di separare le responsabilità nel progetto.  
La classe madre `DatasetArticoliModel` fornisce le funzionalità di base legate alla struttura del dataset, mentre le classi figlie estendono questa logica per introdurre operazioni specifiche:  

- `RicercaUnitariaModel`: gestisce ricerche base basate su un singolo criterio.
- `RicercaCombinataModel`: supporta ricerche avanzate che combinano più criteri contemporaneamente.
- `VisualizzazioneModel`: gestisce la presentazione ordinata dei risultati delle ricerche.
- `RegistrazioneModel`: permette l’inserimento guidato di nuovi articoli all'interno del dataset.
- `AnalisiModel`: analizza la distribuzione degli articoli e fornisce riepiloghi numerici.

Questo gruppo di classi prende il nome "Model" perché rappresenta la logica applicativa del sistema, distinta dall’interfaccia utente (View) e dalla gestione del flusso di controllo (Controller).  

#### Classe `DatasetArticoliModel` - base.

```python
class DatasetArticoliModel:
    def __init__(self, dataset_articoli: dict) -> None:
        """
        Classe Model base per la gestione, elaborazione e rappresentazione strutturata di dataset di articoli.
        
        Fornisce un'astrazione dati centralizzata che gestisce la struttura del dataset,
        metodi helper per l'elaborazione interna, operazioni di ordinamento e filtraggio,
        e strumenti per la visualizzazione tabellare formattata.
        
        La classe implementa il pattern Model del MVC, incapsulando la logica dati e le
        trasformazioni, ed è progettata per essere estesa da classi specializzate che
        implementano funzionalità specifiche come ricerca, analisi o registrazione.
        
        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
                
        Note:
            - Il Model gestisce esclusivamente la logica dati, mantenendo la separazione
              delle responsabilità dal Controller (logica applicativa) e dalla View (presentazione).
            - Tutti gli attributi dinamici sono calcolati automaticamente dal dataset principale
              e vengono mantenuti coerenti attraverso il metodo '_aggiorna_strutture_interne'.
        """
        # COSTANTI - Etichette standardizzate per le colonne della tabella (str).
        # Definizioni utilizzate in tutta l'applicazione per garantire consistenza
        # nelle visualizzazioni tabellari.
        self._COLONNA_POS = "Pos."
        self._COLONNA_TITOLO = "Titolo"
        self._COLONNA_CATEGORIA = "Categoria"
        self._COLONNA_NARRATIVA = "Tipo di narrativa"

        # ATTRIBUTI DINAMICI - struttura dati principale del dataset di articoli.
        # Mantiene la struttura originale a dizionario nidificato per compatibilità
        # con le operazioni di ricerca e aggiornamento.
        self._dataset_articoli = dataset_articoli
 
        # list[tuple[int, str, str, str]] - Lista appiattita di tutti gli articoli.
        # Trasforma la struttura nidificata in una lista piatta di tuple per operazioni
        # efficienti di ordinamento, filtraggio e iterazione completa.
        # Formato: (Posizionamento, Categoria, TipoNarrativa, Titolo)
        self._articoli = [
            (articolo['Posizionamento'], categoria, articolo['TipoNarrativa'], articolo['Titolo'])
            for categoria, articoli in self._dataset_articoli.items()
            for articolo in articoli
        ]
        
        # list[tuple[int, str, str, str]] - Articoli ordinati per posizionamento crescente.
        # Vista pre-calcolata del dataset ordinata numericamente per facilitare
        # operazioni di range, ricerca per posizione e analisi sequenziali.
        # Utilizza il metodo helper '_get_key_estrazione_valore' per specificare
        # il criterio di ordinamento (indice 0 = posizionamento).
        # Formato: (posizione, categoria, narrativa, titolo)
        self._articoli_per_pos_crescente = sorted(
            self._articoli,
            key=self._get_key_estrazione_valore(0)  # 0 = posizionamento.
        )
 
        # list[int] - Lista dei soli posizionamenti, estratti dagli articoli ordinati.
        # Utilizzata per calcoli statistici, determinazione di valori estremi
        # e validazioni di range.
        self._pos_articoli = [articolo[0] for articolo in self._articoli_per_pos_crescente]

        # int - Calcolo automatico dei valori estremi e prossima posizione disponibile.
        # Questi valori sono fondamentali per le operazioni di inserimento nuovo articolo
        # e per le validazioni di input dell'utente.
        self._pos_max = max(self._pos_articoli)        # Posizionamento massimo attualmente presente.
        self._pos_nuovo = self._pos_max + 1           # Prossima posizione disponibile per inserimento.
        
    # ======================== METODI ========================
    def _aggiorna_strutture_interne(self) -> None:
        """
        Sincronizza e aggiorna tutti gli attributi dinamici derivati dopo modifiche al dataset principale.
        
        Questo metodo deve essere invocato ogni volta che il dataset principale
        '_dataset_articoli' viene modificato (ad esempio dopo l'inserimento di un
        nuovo articolo o la modifica di uno esistente) per garantire la coerenza
        tra la struttura dati principale e tutte le viste derivate.
        
        Il metodo ricalcola completamente le seguenti strutture dati derivate:
            - '_articoli': lista appiattita di tutti gli articoli,
            - '_articoli_per_pos_crescente': articoli ordinati per posizione crescente,
            - '_pos_articoli': lista estratta dei soli posizionamenti,
            - '_pos_max': valore massimo di posizionamento presente,
            - '_pos_nuovo': prossima posizione disponibile per nuovo inserimento.
            
        Note:
            - L'aggiornamento è necessario perché gli attributi derivati sono
              calcolati solo all'inizializzazione e non si aggiornano automaticamente
              quando il dataset principale cambia.
            - L'operazione garantisce la consistenza dei dati in tutte le viste e operazioni successive.
            - Utilizza il metodo helper interno '_get_key_estrazione_valore' per
              specificare il criterio di ordinamento (per posizionamento) alla funzione `sorted`.
        """
        # Ricalcola la lista appiattita da zero per riflettere tutte le modifiche.
        self._articoli = [
            (articolo['Posizionamento'], categoria, articolo['TipoNarrativa'], articolo['Titolo'])
            for categoria, articoli in self._dataset_articoli.items()
            for articolo in articoli
        ]

        # Riordina la lista appiattita per posizionamento crescente.
        self._articoli_per_pos_crescente = sorted(
            self._articoli,
            key=self._get_key_estrazione_valore(0)
        )

        # Estrai la lista aggiornata dei posizionamenti.
        self._pos_articoli = [articolo[0] for articolo in self._articoli_per_pos_crescente]

        # Ricalcola i valori estremi e la prossima posizione disponibile.
        self._pos_max = max(self._pos_articoli)
        self._pos_nuovo = self._pos_max + 1
        
    def _stampa_tabella(
        self, 
        titolo: str, 
        colonne: list[tuple[str, str, bool]], 
        righe: list[tuple[str, ...]]
    ) -> None:
        """
        Costruisce e stampa una tabella formattata a schermo con layout professionale.
        
        Wrapper di convenienza che combina la creazione della tabella con la visualizzazione immediata.
        
        Args:
            titolo (str): titolo descrittivo da mostrare sopra la tabella,
                         fornisce contesto immediato sul contenuto della tabella.
            colonne (list[tuple[str, str, bool]]): configurazione delle colonne della tabella.
                Ogni tupla contiene:
                - nome_colonna (str): intestazione della colonna,
                - allineamento (str): 'left', 'center' o 'right',
                - no_wrap (bool): se True, impedisce al testo di andare a capo automaticamente,
                                  utile per colonne con contenuto breve.
            righe (list[tuple[str, ...]]): dati da visualizzare nella tabella.
                Ogni tupla rappresenta una riga completa, con valori nell'ordine
                corrispondente alle colonne definite.
        
        Note:
            - Utilizza il metodo helper interno '_crea_tabella_generica' per la
              costruzione effettiva dell'oggetto Table.
            - La stampa viene eseguita tramite la console di Rich, che garantisce
              formattazione avanzata e supporto a colori e stili.
            - Tutti i valori nelle righe vengono convertiti automaticamente a stringa
              per garantire compatibilità con la libreria Rich.
        """
        tabella = self._crea_tabella_generica(titolo, colonne, righe)
        console.print(tabella)
                
    # ======================== METODI HELPER ========================
    
    def _get_key_estrazione_valore(self, indice: int) -> callable:
        """
        Crea e restituisce una funzione che estrae un elemento specifico da una tupla,
        progettata per essere utilizzata come parametro 'key' nelle funzioni di
        ordinamento come `sorted()`, `sort()`, `min()`, `max()`.
        
        Args:
            indice (int): posizione dell'elemento da estrarre dalle tuple.
                        
        Returns:
            callable: funzione che accetta una tupla e restituisce l'elemento
                     all'indice specificato, pronta per essere utilizzata come
                     parametro 'key' nelle funzioni di ordinamento.
            
        Example:
            >>> articoli = [('Notizia A', 'Sport', 3), ('Notizia B', 'Cronaca', 1), ('Notizia C', 'Sport', 2)]
            
            >>> # Ordina per categoria (indice 1)
            >>> per_categoria = self._get_key_estrazione_valore(1)
            >>> sorted(articoli, key=per_categoria)
                [('Notizia B', 'Cronaca', 1), ('Notizia A', 'Sport', 3), ('Notizia C', 'Sport', 2)]
            
            >>> # Ordina per posizionamento (indice 2)  
            >>> per_posizione = self._get_key_estrazione_valore(2)
            >>> sorted(articoli, key=per_posizione)
                [('Notizia B', 'Cronaca', 1), ('Notizia C', 'Sport', 2), ('Notizia A', 'Sport', 3)]
        """
        def key_estrazione(tupla: tuple) -> any:
            """Funzione interna che estrae l'elemento all'indice specificato."""
            return tupla[indice]
        return key_estrazione

    def _crea_tabella_generica(
        self, 
        titolo: str, 
        colonne: list[tuple[str, str, bool]], 
        righe: list[tuple[str, ...]], 
        show_header: bool = True
    ) -> Table:
        """
        Crea e configura un oggetto Table della libreria Rich per visualizzazione avanzata.
        
        Costruisce una tabella formattata con titolo, intestazioni colonne e righe dati,
        applicando le impostazioni di allineamento e wrapping specificate.
        
        Args:
            titolo (str): titolo della tabella, visualizzato in stile prominente.
            colonne (list[tuple[str, str, bool]]): specifica di configurazione per ogni colonna.
                Ogni tupla contiene:
                - nome_colonna (str): testo dell'intestazione della colonna,
                - allineamento (str): allineamento del testo nella colonna
                                     ('left', 'center', o 'right'),
                - no_wrap (bool): se True, impedisce al testo di andare a capo
                                 automaticamente, mantenendo righe compatte.
            righe (list[tuple[str, ...]]): dati da inserire nella tabella.
                Ogni tupla rappresenta una riga, con valori corrispondenti
                all'ordine delle colonne definite.
            show_header (bool): se True mostra l'intestazione delle colonne,
                               se False crea una tabella senza intestazioni.

        Returns: 
            Table: oggetto Table della libreria Rich completamente configurato,
                  pronto per la visualizzazione tramite `console.print()`.
        
        Note:
            - Tutti i valori nelle righe vengono convertiti esplicitamente a stringa
              tramite `map(str, riga)` per garantire compatibilità con l'API Rich.
            - La configurazione delle colonne viene applicata sequenzialmente nell'ordine
              fornito, determinando l'ordine di visualizzazione da sinistra a destra.
            - L'oggetto Table risultante può essere ulteriormente personalizzato
              (colori, bordi, stili) prima della visualizzazione finale.
        """
        # Crea l'oggetto Table base con titolo e opzione di visualizzazione intestazione.
        tabella = Table(title=titolo, show_header=show_header)
        
        # Aggiunge tutte le colonne specificate con le loro configurazioni.
        for nome_colonna, allineamento, wrap in colonne:
            tabella.add_column(
                nome_colonna, 
                justify=allineamento, 
                no_wrap=wrap)
        
        # Aggiunge tutte le righe di dati, convertendo ogni valore a stringa.
        for riga in righe:
            tabella.add_row(*map(str, riga))

        return tabella
```

```python
class RicercaUnitariaModel(DatasetArticoliModel):
    def __init__(self, dataset_articoli: dict):
        """        
        Classe Model specializzata per operazioni di ricerca di articoli basate su un singolo criterio.
        
        Estende 'DatasetArticoliModel' fornendo implementazioni per le ricerche unitarie 
        che filtrano gli articoli a seconda del criterio singolo selezionato:
        - Per categoria specifica,
        - Per posizionamento singolo,
        - Per tipo di narrativa,
        - Per parola chiave nel titolo.
        
        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
        
        Note:
            - Il Model si occupa esclusivamente della logica di ricerca e filtraggio,
              delegando la formattazione e visualizzazione ai metodi ereditati dalla
              classe base.
        """
        super().__init__(dataset_articoli)
    
    # ======================== METODI DI RICERCA SINGOLA ========================
    def _ricerca_per_categoria(self, categoria_scelta: str) -> None:
        """
        Filtra e visualizza tutti gli articoli appartenenti a una specifica categoria.
        
        Accede direttamente al dataset usando il nome della categoria come chiave,
        estrae tutti gli articoli di quella categoria e li visualizza in una tabella
        con posizionamento, tipo di narrativa e titolo.
        
        Args:
            categoria_scelta (str): nome della categoria da cercare (case-sensitive,
                                   deve corrispondere esattamente alle chiavi del
                                   dizionario '_dataset_articoli').
            
        Note: 
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
            - Utilizza i seguenti attributi della classe base:
              * '_dataset_articoli' (dict): struttura dati principale contenente tutti
                gli articoli organizzati per categoria.
              * '_COLONNA_POS', '_COLONNA_NARRATIVA', '_COLONNA_TITOLO' (str): costanti
                per le etichette delle colonne della tabella.
            - La ricerca restituisce tutti gli articoli della categoria, mantenendo
              l'ordine originale presente nel dataset.
        """
        # Estrae tutti gli articoli della categoria specificata.
        # list[tuple[int, str, str]] - Formato righe: (posizionamento, tipo_narrativa, titolo).
        righe = [
            [articolo['Posizionamento'], articolo['TipoNarrativa'], articolo['Titolo']]
            for articolo in self._dataset_articoli.get(categoria_scelta, [])
        ]

        # Configurazione colonne per visualizzazione ottimizzata per ricerca per categoria.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_POS, "center", False),      # Posizionamento centrato, testo a capo automatico.
            (self._COLONNA_NARRATIVA, "right", True),  # Narrativa allineata a destra, testo su singola riga.
            (self._COLONNA_TITOLO, "left", True)       # Titolo allineato a sinistra, testo su singola riga.
        ]


        self._stampa_tabella(f"Articoli nella categoria '{categoria_scelta.capitalize()}'.", colonne, righe)

    def _ricerca_per_pos_singolo(self, pos_scelto: str) -> None:
        """
        Cerca e visualizza l'articolo corrispondente a un specifico posizionamento.
        
        Cerca in tutte le categorie l'articolo che ha esattamente il posizionamento
        indicato. Poiché ogni articolo ha un posizionamento unico, questa ricerca
        trova al massimo un risultato.
        
        Args:
            pos_scelto (str): posizionamento da cercare (verrà convertito a int
                             internamente). Deve corrispondere a un valore
                             presente nel dataset.
            
        Note:
            - Il posizionamento è garantito essere unico nel dataset, 
              quindi la ricerca restituisce al massimo un risultato.
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
            - Utilizza i seguenti attributi della classe base:
              * '_dataset_articoli' (dict): struttura dati principale per la scansione.
              * '_COLONNA_CATEGORIA', '_COLONNA_NARRATIVA', '_COLONNA_TITOLO' (str):
                costanti per le etichette delle colonne.
            - Se nessun articolo corrisponde al posizionamento, viene visualizzata
              una tabella vuota.
        """
        # Scansione completa di tutte le categorie per trovare l'articolo con il posizionamento specificato.
        # list[tuple[str, str, str]] - Formato righe: (categoria, narrativa, titolo).
        righe = [] 
        for categoria, articoli in self._dataset_articoli.items():
            for articolo in articoli:
                if articolo['Posizionamento'] == int(pos_scelto):
                    righe.append([
                        categoria.capitalize(),
                        articolo['TipoNarrativa'],
                        articolo['Titolo']
                    ])

        # Configurazione colonne per visualizzazione ottimizzata per risultato singolo.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_CATEGORIA, "center", False), # Categoria centrata, testo a capo automatico.
            (self._COLONNA_NARRATIVA, "center", True),  # Narrativa centrata, testo su singola riga.
            (self._COLONNA_TITOLO, "center", True)      # Titolo centrato, testo su singola riga.
        ]
        
        self._stampa_tabella(f"Articoli con posizionamento {pos_scelto}.", colonne, righe)

    def _ricerca_per_narrativa(self, narrativa_scelta: str) -> None:
        """
        Filtra e visualizza tutti gli articoli di un specifico tipo di narrativa.
        
        Cerca in tutte le categorie gli articoli che hanno il tipo di narrativa
        indicato. Mostra dove appare questa narrativa nelle diverse categorie.
        
        Args:
            narrativa_scelta (str): nome della narrativa da cercare (case-sensitive,
                                   deve corrispondere esattamente ai valori del campo
                                   'TipoNarrativa' negli articoli).
            
        Note:
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
            - Utilizza i seguenti attributi della classe base:
              * '_dataset_articoli' (dict): struttura dati principale per la scansione.
              * '_COLONNA_CATEGORIA', '_COLONNA_POS', '_COLONNA_TITOLO' (str):
                costanti per le etichette delle colonne.
        """
        # Scansione completa di tutti gli articoli per identificare quelli con la narrativa specificata.
        # list[tuple[str, int, str]] - Formato righe: (categoria, posizionamento, titolo).
        righe = [] 
        for categoria, articoli in self._dataset_articoli.items():
            for articolo in articoli:
                if narrativa_scelta in articolo['TipoNarrativa']:
                    righe.append([
                        categoria.capitalize(),
                        articolo['Posizionamento'],
                        articolo['Titolo']
                    ])

        # Configurazione colonne per visualizzazione ottimizzata per ricerca per narrativa.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_CATEGORIA, "right", False), # Categoria allineata a destra, testo a capo automatico.
            (self._COLONNA_POS, "center", True),       # Posizionamento centrato, testo su singola riga.      
            (self._COLONNA_TITOLO, "left", True)       # Titolo allineato a sinistra, testo su singola riga.
        ]

        self._stampa_tabella(f"Articoli appartenenti alla narrativa '{narrativa_scelta}'.", colonne, righe)

    def _ricerca_per_parola_nei_titoli(self, parola_scelta: str) -> None:
        """
        Cerca e visualizza articoli contenenti una parola specifica nei loro titoli.
        
        Cerca in tutti i titoli degli articoli quelli che contengono la parola
        indicata, senza distinguere tra maiuscole e minuscole. La parola può
        apparire in qualsiasi parte del titolo.
        
        Args:
            parola_scelta (str): parola da cercare nei titoli (ricerca case-insensitive,
                                corrispondenza parziale supportata).
            
        Note:
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
            - Utilizza i seguenti attributi della classe base:
              * '_dataset_articoli' (dict): struttura dati principale per la scansione.
              * '_COLONNA_CATEGORIA', '_COLONNA_POS', '_COLONNA_NARRATIVA',
                '_COLONNA_TITOLO' (str): costanti per le etichette delle colonne.
            - La ricerca è case-insensitive: converte sia la parola cercata che i
              titoli in lowercase per il confronto.
            - Supporta la corrispondenza parziale: la parola può apparire in qualsiasi
              posizione all'interno del titolo.
            - Se nessun titolo contiene la parola, viene visualizzata una tabella vuota.
        """
        # Ricerca case-insensitive con scansione completa di tutti i titoli.
        # list[tuple[str, int, str, str]] - Formato righe: (categoria, posizionamento, narrativa, titolo).
        righe = [] 
        for categoria, articoli in self._dataset_articoli.items():
            for articolo in articoli:
                if parola_scelta.lower() in articolo['Titolo'].lower():
                    righe.append([
                        categoria.capitalize(),
                        articolo['Posizionamento'],
                        articolo['TipoNarrativa'],
                        articolo['Titolo']
                    ])

        # Configurazione colonne per visualizzazione completa dei risultati di ricerca testuale.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_CATEGORIA, "right", False), # Categoria allineata a destra, testo a capo automatico.
            (self._COLONNA_POS, "center", True),       # Posizionamento centrato, testo su singola riga.
            (self._COLONNA_NARRATIVA, "right", True),  # Narrativa allineata a destra, testo su singola riga.
            (self._COLONNA_TITOLO, "left", True)       # Titolo allineato a sinistra, testo su singola riga.
        ]
        
        self._stampa_tabella(f"Articoli con titoli contenenti la parola '{parola_scelta}'.", colonne, righe)
```

#### Classe `RicercaCombinataModel` - derivata.

```python
class RicercaCombinataModel(DatasetArticoliModel):
    def __init__(self, dataset_articoli: dict):
        """
        Classe Model specializzata per operazioni di ricerca di articoli basate su criteri multipli.
        
        Estende 'DatasetArticoliModel' fornendo implementazioni per ricerche avanzate
        che filtrano gli articoli a seconda dei criteri multipli selezionati:
        - Range di posizionamenti,
        - Categoria + Tipo di narrativa,
        - Categoria + Parola nel titolo,
        - Narrativa + Parola nel titolo.
        
        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
                
        Note:
            - Il Model si occupa esclusivamente della logica di ricerca e filtraggio combinato,
              delegando la formattazione e visualizzazione ai metodi ereditati dalla classe base.
            - Tutte le ricerche operano con AND logico: un articolo deve soddisfare tutti i criteri
              specificati per essere incluso nei risultati.
        """
        super().__init__(dataset_articoli)

    # ======================== METODI DI RICERCA COMBINATA ========================
    
    def _ricerca_per_range_pos(self, pos_min_scelto: str, pos_max_scelto: str) -> None:   
        """
        Filtra e visualizza articoli compresi in un intervallo di posizionamenti.
        
        Cerca tutti gli articoli provvisti di un posizionamento compreso nel range selezionato,
        compreso tra un valore minimo e massimo inclusivi.
        Sfrutta la lista pre-ordinata '_articoli_per_pos_crescente' per efficienza,
        evitando di dover riordinare i risultati.
        
        Args:
            pos_min_scelto (str): estremo inferiore dell'intervallo (inclusivo).
                                Verrà convertito internamente a intero.
            pos_max_scelto (str): estremo superiore dell'intervallo (inclusivo).
                                Verrà convertito internamente a intero.
            
        Note:
            - L'intervallo include entrambi gli estremi: [min, max] (inclusivo-inclusivo).
            - Gli articoli sono mostrati in ordine di posizionamento crescente,
              sfruttando l'ordinamento preesistente della lista sorgente.
            - Se l'intervallo non contiene articoli, viene visualizzata una tabella vuota.
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
            - Utilizza i seguenti attributi della classe base:
              * '_articoli_per_pos_crescente' (list[tuple[int, str, str, str]]): lista
                pre-ordinata per posizionamento crescente che ottimizza la ricerca per range.
              * '_COLONNA_POS', '_COLONNA_CATEGORIA', '_COLONNA_NARRATIVA',
                '_COLONNA_TITOLO' (str): costanti per le etichette delle colonne.
        """     
        # Genera l'intervallo numerico completo tra i due estremi inclusivi.
        # list[int] - intervallo di posizionamenti da pos_min_scelto a pos_max_scelto (inclusi).
        range_pos = list(range(int(pos_min_scelto), int(pos_max_scelto) + 1))
        
        # Filtra gli articoli nell'intervallo utilizzando la lista pre-ordinata per efficienza.
        # list[tuple[int, str, str, str]] - Formato righe: (posizione, categoria, narrativa, titolo).
        righe = [
            [posizione, categoria.capitalize(), narrativa, titolo]
            for posizione, categoria, narrativa, titolo in self._articoli_per_pos_crescente
            if posizione in range_pos
        ]
        
         # Configurazione colonne ottimizzata per visualizzazione di range di posizionamenti.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_POS, "center", True),      # Posizionamento centrato, testo su singola riga.
            (self._COLONNA_CATEGORIA, "right", True), # Categoria allineata a destra, testo su singola riga.
            (self._COLONNA_NARRATIVA, "right", True), # Narrativa allineata a destra, testo su singola riga.
            (self._COLONNA_TITOLO, "left", True)      # Titolo allineato a sinistra, testo su singola riga.
        ]

        self._stampa_tabella(
            f"Articoli nel range di posizionamenti da {pos_min_scelto} a {pos_max_scelto}.", 
            colonne, righe)
        
    def _ricerca_per_categoria_e_narrativa(self, categoria_scelta: str, narrativa_scelta: str) -> None:
        """
        Filtra e visualizza articoli che appartengono sia a una categoria specifica che a un tipo di narrativa specifico.
        
        Cerca tutti gli articoli che appartengono sia a una categoria specifica che a un tipo di narrativa specifico.
        Applica due filtri contemporaneamente: l'appartenenza esatta alla categoria e l'appartenenza esatta alla narrativa.
        Sfrutta l'accesso diretto alla categoria nel dataset per efficienza,
        limitando la ricerca solo agli articoli di quella categoria prima di applicare il filtro per narrativa.
        
        Args:
            categoria_scelta (str): categoria specifica da cercare (case-sensitive,
                                   deve corrispondere esattamente alle chiavi del dizionario).
            narrativa_scelta (str): narrativa specifica da cercare (case-sensitive,
                                   deve corrispondere esattamente ai valori del campo 'TipoNarrativa').
            
        Note:
            - Entrambi i criteri devono essere soddisfatti contemporaneamente (AND logico).
            - La ricerca è molto selettiva e tipicamente restituisce un numero limitato di risultati.
            - Se nessun articolo soddisfa entrambi i criteri, viene visualizzata una tabella vuota.
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
            - Utilizza i seguenti attributi della classe base:
              * '_dataset_articoli' (dict): struttura dati principale per la ricerca diretta.
              * '_COLONNA_POS', '_COLONNA_TITOLO' (str): costanti per le etichette delle colonne.
            - La visualizzazione include solo posizionamento e titolo poiché categoria
              e narrativa sono note dal contesto di ricerca.
        """     
        # list[tuple[int, str]] - Formato righe: (posizionamento, titolo).
        righe = [] 
        
        # Cerca solo nella categoria specificata articoli con la narrativa esatta.
        # La ricerca è ottimizzata: accede direttamente alla categoria senza scansionare tutto il dataset.
        for categoria, articoli in self._dataset_articoli.items():
            if categoria == categoria_scelta:
                for articolo in articoli:
                    if  articolo['TipoNarrativa'] == narrativa_scelta:
                        righe.append([
                            articolo['Posizionamento'],
                            articolo['Titolo']
                        ])
            
        # Configurazione colonne minimalista per risultati molto specifici.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).          
        colonne = [
            (self._COLONNA_POS, "center", True),       # Posizionamento centrato, testo su singola riga.
            (self._COLONNA_TITOLO, "left", True)       # Titolo allineato a sinistra, testo su singola riga.
        ]
        
        self._stampa_tabella(
            f"Articoli nella categoria '{categoria_scelta.capitalize()}'\ncon tipo di narrativa '{narrativa_scelta}'.", 
            colonne, righe)
            
    def _ricerca_per_categoria_e_parola_nei_titoli(self, categoria_scelta: str, parola_scelta: str) -> None:
        """
        Filtra e visualizza articoli di una categoria specifica che contengono una parola specifica nei titoli.
        
        Cerca tutti gli articoli di una categoria specifica che contengono una parola nel titolo.
        Combina due condizioni: l'appartenenza alla categoria e la presenza della parola nei titoli,
        senza distinguere tra maiuscole e minuscole.
        Sfrutta l'accesso diretto alla categoria nel dataset per efficienza,
        limitando la ricerca solo agli articoli di quella categoria.
        
        Args:
            categoria_scelta (str): categoria specifica da cercare (case-sensitive,
                                   deve corrispondere esattamente a una chiave del dizionario).
            parola_scelta (str): parola da cercare nei titoli (case-insensitive,
                                corrispondenza parziale supportata).
            
        Note:
            - Entrambi i criteri devono essere soddisfatti contemporaneamente (AND logico).
            - La ricerca testuale è case-insensitive: non distingue tra maiuscole e minuscole.
            - La parola può apparire in qualsiasi posizione del titolo.
            - Se nessun articolo soddisfa entrambi i criteri, viene visualizzata una tabella vuota.
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
            - Utilizza i seguenti attributi della classe base:
              * '_dataset_articoli' (dict): struttura dati principale per la ricerca diretta.
              * '_COLONNA_POS', '_COLONNA_NARRATIVA', '_COLONNA_TITOLO' (str): costanti
                per le etichette delle colonne.
        """  
        # list[tuple[int, str, str]] - Formato righe: (posizionamento, narrativa, titolo).
        righe = [] 
        
        # Cerca solo nella categoria specificata articoli con titoli contenenti la parola chiave.
        # La ricerca è ottimizzata: accede direttamente alla categoria senza scansionare tutto il dataset.
        for categoria, articoli in self._dataset_articoli.items():
            if categoria == categoria_scelta:
                for articolo in articoli:
                    if parola_scelta.lower() in articolo['Titolo'].lower():
                        righe.append([
                            articolo['Posizionamento'],
                            articolo['TipoNarrativa'],
                            articolo['Titolo']
                        ])
        
        # Configurazione colonne per visualizzare il contesto completo dei risultati.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).            
        colonne = [
            (self._COLONNA_POS, "center", True),      # Posizionamento centrato, testo su singola riga.
            (self._COLONNA_NARRATIVA, "right", True), # Narrativa allineata a destra, testo su singola riga.
            (self._COLONNA_TITOLO, "left", True)      # Titolo allineato a sinistra, testo su singola riga.
        ]
        
        self._stampa_tabella(
            f"Articoli nella categoria '{categoria_scelta.capitalize()}'\ncon titoli contenenti la parola '{parola_scelta}'.", 
            colonne, righe)
        
    def _ricerca_per_narrativa_e_parola_nei_titoli(self, narrativa_scelta: str, parola_scelta: str) -> None:
        """
        Filtra e visualizza articoli di un tipo specifico di narrativa che contengono una parola specifica nei titoli.
        
        Cerca tutti gli articoli di un tipo di narrativa specifico che contengono una parola nel titolo.
        Combina due condizioni: l'appartenenza alla narrativa e la presenza della parola nei titoli,
        senza distinguere tra maiuscole e minuscole.
        Scansiona tutto il dataset per trovare la narrativa in tutte le categorie,
        applicando prima il filtro per narrativa per efficienza.
        
        Args:
            narrativa_scelta (str): narrativa specifica da cercare (case-sensitive,
                                   deve corrispondere esattamente ai valori del campo 'TipoNarrativa').
            parola_scelta (str): parola da cercare nei titoli (case-insensitive).
            
        Note:
            - Entrambi i criteri devono essere soddisfatti contemporaneamente (AND logico).
            - La ricerca testuale è case-insensitive: non distingue tra maiuscole e minuscole.
            - La corrispondenza è parziale: la parola può apparire in qualsiasi posizione del titolo.
            - Poiché la narrativa può apparire in diverse categorie, la ricerca scansiona
              tutto il dataset ma filtra prima per narrativa per efficienza.
            - Se nessun articolo soddisfa entrambi i criteri, viene visualizzata una tabella vuota.
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
            - Utilizza i seguenti attributi della classe base:
              * '_dataset_articoli' (dict): struttura dati principale per la scansione completa.
              * '_COLONNA_CATEGORIA', '_COLONNA_POS', '_COLONNA_TITOLO' (str): costanti
                per le etichette delle colonne.
        """ 
        # list[tuple[str, int, str]] - Formato righe: (categoria, posizionamento, titolo).
        righe = [] 
        
        # Scansiona tutto il dataset per articoli della narrativa specificata con titoli contenenti la parola.
        # L'ordine dei filtri è ottimizzato: prima verifica la narrativa (più veloce), poi la ricerca testuale.
        for categoria, articoli in self._dataset_articoli.items():
            for articolo in articoli:
                if  articolo['TipoNarrativa'] == narrativa_scelta:
                    if parola_scelta.lower() in articolo['Titolo'].lower():
                        righe.append([
                            categoria.capitalize(),
                            articolo['Posizionamento'],
                            articolo['Titolo']
                        ])
               
        # Configurazione colonne per mostrare la distribuzione trasversale della narrativa.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).         
        colonne = [
            (self._COLONNA_CATEGORIA, "right", False), # Categoria allineata a destra, testo a capo automatico.
            (self._COLONNA_POS, "center", True),       # Posizionamento centrato, testo su singola riga.
            (self._COLONNA_TITOLO, "left", True)       # Titolo allineato a sinistra, testo su singola riga.
        ]
        
        self._stampa_tabella(
            f"Articoli appartenenti alla narrativa '{narrativa_scelta}'\ncon titoli contenenti la parola '{parola_scelta}'.", 
            colonne, righe)
```

#### Classe `VisualizzazioneModel` - derivata.

```python
class VisualizzazioneModel(DatasetArticoliModel):
    def __init__(self, dataset_articoli: dict):
        """
        Classe Model specializzata per la visualizzazione completa e organizzata del dataset articoli.
        
        Estende 'DatasetArticoliModel' fornendo implementazioni per la presentazione
        strutturata di tutti gli articoli secondo diversi criteri:
        - Raggruppati per categoria,
        - Raggruppati per tipo di narrativa,
        - Ordinati per posizionamento crescente.
        
        Riutilizza le funzionalità di ricerca già implementate in 'RicercaUnitariaModel', 
        evitando duplicazione di codice e mantenendo coerenza nelle visualizzazioni 
        tra ricerca e esplorazione.
        
        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
                
        Note:
            - Il Model si occupa esclusivamente della logica di organizzazione e raggruppamento
              per la visualizzazione completa, delegando la presentazione tabellare effettiva
              ai metodi ereditati dalla classe base o composti da 'RicercaUnitariaModel'.
            - L'uso della composizione con 'RicercaUnitariaModel' garantisce che le visualizzazioni
              utilizzino esattamente lo stesso formato e layout delle corrispondenti ricerche.
        """
        super().__init__(dataset_articoli)
        
        # RicercaUnitariaModel - Composizione per riutilizzare le funzionalità di ricerca.
        # Invece di reimplementare i metodi di visualizzazione per categoria e narrativa,
        # questa classe delega a 'RicercaUnitariaModel' che già contiene implementazioni
        # ottimizzate e testate per quelle funzionalità.
        self._ricerca_unitaria = RicercaUnitariaModel(dataset_articoli)
        
    # ======================== METODI DI VISUALIZZAZIONE ========================
    def _visualizza_dataset_per_categoria(self) -> None:
        """
        Visualizza tutti gli articoli organizzati in gruppi per categoria.
        
        Presenta il dataset completo suddividendolo in tabelle separate per ogni
        categoria esistente, mostrando tutte le categorie nell'ordine originale
        in cui compaiono nel dataset. Ogni tabella categoria riutilizza esattamente
        lo stesso formato e layout della ricerca per categoria singola.
        
        Note:
            - Mostra tutte le categorie presenti nel dataset, senza esclusioni.
            - Le categorie sono presentate nell'ordine originale di apparizione nel datase.
            - Ogni categoria viene visualizzata in una tabella distinta con titolo appropriato.
            - La visualizzazione è identica a eseguire una ricerca per categoria su ciascuna categoria.
            - Utilizza il metodo '_ricerca_per_categoria' (metodo) della classe composta 
              'RicercaUnitariaModel' per la visualizzazione di ogni singola categoria.
            - Utilizza '_dataset_articoli' (dict): attributo ereditato dalla classe
              base che contiene la struttura dati principale per iterare sulle categorie.
        """
        # Itera sequenzialmente su tutte le categorie presenti nel dataset
        # e visualizza gli articoli di ciascuna categoria in una tabella separata.
        # Equivale a: for categoria in self._dataset_articoli.keys()
        for categoria in self._dataset_articoli:
            self._ricerca_unitaria._ricerca_per_categoria(categoria)

    def _visualizza_dataset_per_narrativa(self) -> None:
        """
        Visualizza tutti gli articoli organizzati in gruppi per tipo di narrativa.
        
        Presenta il dataset completo suddividendolo in tabelle separate per ogni
        tipo di narrativa unico presente nel dataset, mostrando le narrative in
        ordine alfabetico, esplorando come ciascun genere narrativo si distribuisce 
        attraverso le diverse categorie tematiche.
        
        Note:
            - Identifica e mostra tutti i tipi di narrativa unici presenti nel dataset.
            - Le narrative sono presentate in ordine alfabetico crescente.
            - Ogni narrativa viene visualizzata in una tabella distinta che mostra
              tutti gli articoli di quel genere narrativo, indipendentemente dalla
              categoria di appartenenza.
            - La visualizzazione è identica a eseguire una ricerca per narrativa
              su ciascun tipo di narrativa.
            - Utilizza il metodo '_ricerca_per_narrativa' (metodo) della classe composta 
              'RicercaUnitariaModel' per la visualizzazione di ogni singola narrativa.
            - Utilizza '_dataset_articoli' (dict): attributo ereditato dalla classe
              base che contiene la struttura dati principale per estrarre le narrative.
        """
        # Estrae tutti i tipi di narrativa unici dal dataset utilizzando
        # una set comprehension per garantire l'unicità ed eliminare duplicati.
        # set(str) - Insieme di tutti i tipi di narrativa distinti.
        narrative = set(
            articolo['TipoNarrativa']
            for articoli in self._dataset_articoli.values()
            for articolo in articoli
        )

        # Converte il set in lista e ordina alfabeticamente per garantire
        # una presentazione consistente e prevedibile delle narrative.
        # list[str] - Narrative ordinate alfabeticamente in ordine crescente.
        narrative_ordinate = sorted(narrative)

        # Visualizza gli articoli per ogni tipo di narrativa in ordine alfabetico,
        # mostrando ciascuna narrativa in una tabella separata.
        for narrativa in narrative_ordinate:
            self._ricerca_unitaria._ricerca_per_narrativa(narrativa)

    def _visualizza_dataset_per_pos_crescente(self) -> None:
        """
        Visualizza tutti gli articoli in ordine sequenziale di posizionamento crescente.
        
        Presenta il dataset completo in una singola tabella unificata con tutti gli
        articoli ordinati numericamente per posizionamento, dalla posizione più bassa
        alla più alta, evidenziando la progressione numerica indipendentemente 
        dalla categoria o narrativa di appartenenza.
        
        Note:
            - Mostra tutti gli articoli in una singola tabella continua.
            - L'ordinamento è puramente numerico basato sul campo 'Posizionamento',
              indipendentemente da altri attributi degli articoli.
            - La tabella include tutte le colonne per fornire contesto completo.
            - Sfrutta la lista pre-calcolata '_articoli_per_pos_crescente' per
              efficienza, evitando di dover riordinare i dati a ogni esecuzione.
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
            - Utilizza i seguenti attributi della classe base:
              * '_articoli_per_pos_crescente' (list[tuple[int, str, str, str]]): lista
                pre-ordinata per posizionamento crescente che fornisce i dati già
                organizzati nella sequenza corretta.
              * '_COLONNA_POS', '_COLONNA_CATEGORIA', '_COLONNA_NARRATIVA',
                '_COLONNA_TITOLO' (str): costanti per le etichette delle colonne
                della tabella.
        """
        # Utilizza direttamente la lista già ordinata degli articoli per posizionamento,
        # convertendo ciascuna tupla in una riga della tabella con formattazione appropriata.
        # list[tuple[int, str, str, str]] - Formato righe: (posizione, categoria, narrativa, titolo).
        righe = [ 
            [posizione, categoria.capitalize(), tipo_narrativa, titolo]
            for posizione, categoria, tipo_narrativa, titolo in self._articoli_per_pos_crescente
        ]

        # Configurazione colonne ottimizzata per visualizzazione sequenziale completa.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [            
            (self._COLONNA_POS, "center", False),      # Posizionamento centrato, testo a capo automatico.
            (self._COLONNA_CATEGORIA, "right", True),  # Categoria allineata a destra, testo su singola riga.
            (self._COLONNA_NARRATIVA, "right", True),  # Narrativa allineata a destra, testo su singola riga.
            (self._COLONNA_TITOLO, "left", True)       # Titolo allineato a sinistra, testo su singola riga.
        ]

        self._stampa_tabella("Articoli ordinati per posizionamento.", colonne, righe)
```

#### Classe `RegistrazioneModel` - derivata.

```python
class RegistrazioneModel(DatasetArticoliModel):
    def __init__(self, dataset_articoli: dict):
        """
        Classe Model specializzata per la registrazione e inserimento di nuovi articoli nel dataset.
        
        Estende 'DatasetArticoliModel' fornendo implementazioni per le operazioni
        di aggiunta di nuovi articoli al dataset, gestendo automaticamente:
        - Assegnazione dei posizionamenti,
        - Inserimento nella struttura dati per categoria,
        - Conferma visiva dell'operazione completata.

        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
                
        Note:
            - Il Model si occupa esclusivamente della logica di inserimento e
              aggiornamento dati, delegando la validazione degli input al Controller
              e la presentazione delle interfacce di input alla View.
            - Tutte le operazioni di modifica sono accompagnate da 
              aggiornamenti completi delle strutture dati.
            - L'assegnazione automatica dei posizionamenti garantisce l'unicità
              e la progressione sequenziale senza richiedere intervento utente.
        """
        super().__init__(dataset_articoli)
        
    # ======================== METODI DI REGISTRAZIONE ========================
    def _registra_articolo(self, categoria_scelta: str, narrativa_scelta: str, titolo_scelto: str) -> None:
        """
        Registra un nuovo articolo nel dataset con i parametri specificati dall'utente.
        
        Implementa il processo completo di inserimento di un nuovo articolo:
        1. Crea un nuovo oggetto articolo con il posizionamento auto-generato,
        2. Inserisce l'articolo nella categoria specificata,
        3. Mostra una conferma visiva con tutti i dettagli dell'articolo registrato,
        4. Ricalcola automaticamente tutti i dati che derivano dal dataset per tenere tutto aggiornato.
        
        Questo metodo rappresenta il punto centrale di modifica del dataset
        e garantisce che tutte le operazioni successive (ricerche, visualizzazioni,
        analisi) operino su dati aggiornati e coerenti.
        
        Args:
            - categoria_scelta (str): categoria in cui inserire il nuovo articolo.
                Deve corrispondere esattamente a una chiave esistente nel dataset.
            - narrativa_scelta (str): tipo di narrativa da assegnare al nuovo articolo.
                Deve corrispondere a uno dei tipi di narrativa supportati.
            - titolo_scelto (str): titolo completo del nuovo articolo.
                Deve rispettare tutti i vincoli di validazione definiti nella View.
            
        Note:
            - Assegna automaticamente il prossimo posizionamento disponibile
              (valore di '_pos_nuovo') garantendo progressione sequenziale e univocità.
            - Aggiorna tutte le strutture dati derivate attraverso il metodo
              '_aggiorna_strutture_interne' per preservare l'integrità dei dati
              in tutte le viste e operazioni successive.
            - Visualizza una conferma tabellare con tutti i dettagli dell'articolo
              registrato per fornire feedback immediato all'utente.

            - Utilizza:
              * '_aggiorna_strutture_interne' (metodo): helper ereditato dalla classe base
                che ricalcola e sincronizza tutte le strutture dati derivate dopo la modifica.
              * '_stampa_tabella' (metodo): helper ereditato dalla classe base per la
                creazione e visualizzazione di tabelle formattate di conferma.
        
            - Utilizza i seguenti attributi della classe base:
              * '_pos_nuovo' (int): prossimo posizionamento disponibile da assegnare
                automaticamente al nuovo articolo (calcolato come max attuale + 1).
              * '_dataset_articoli' (dict): struttura dati principale che viene
                modificata con l'aggiunta del nuovo articolo.
              * '_COLONNA_CATEGORIA', '_COLONNA_POS', '_COLONNA_NARRATIVA',
                '_COLONNA_TITOLO' (str): costanti per le etichette delle colonne
                nella tabella di conferma.
        """
        # Crea il dizionario rappresentante il nuovo articolo utilizzando
        # il prossimo posizionamento disponibile calcolato automaticamente.
        # dict{'Posizionamento': int, 'TipoNarrativa': str, 'Titolo': str}
        nuovo_articolo = {
            "Posizionamento": self._pos_nuovo,        # Posizionamento auto-generato sequenziale
            "TipoNarrativa": narrativa_scelta,        # Tipo di narrativa specificato dall'utente
            "Titolo": titolo_scelto                   # Titolo validato specificato dall'utente
        }

        # Aggiunge il nuovo articolo alla lista della categoria specificata
        # all'interno della struttura gerarchica principale del dataset.
        # L'operazione modifica direttamente il dataset in memoria.
        self._dataset_articoli[categoria_scelta].append(nuovo_articolo)

        # Prepara i dati per la tabella di conferma che mostra all'utente
        # tutti i dettagli dell'articolo appena registrato.
        # list[tuple[str, int, str, str]] - Formato righe: (categoria, posizionamento, narrativa, titolo)
        righe = [[categoria_scelta, str(self._pos_nuovo), narrativa_scelta, titolo_scelto]]

        # Configurazione colonne per la tabella di conferma della registrazione.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_CATEGORIA, "center", True), # Categoria centrata, testo su singola riga.
            (self._COLONNA_POS, "center", False),      # Posizionamento centrato, testo a capo automatico.
            (self._COLONNA_NARRATIVA, "center", True), # Narrativa centrata, testo su singola riga.
            (self._COLONNA_TITOLO, "center", True)     # Titolo centrato, testo su singola riga.
        ]

        # Visualizza la conferma della registrazione in formato tabellare formattato,
        # fornendo all'utente feedback immediato e verificabile dell'operazione.
        self._stampa_tabella("Nuovo articolo registrato.", colonne, righe)

        # Aggiorna tutte le strutture dati derivate per riflettere le modifiche
        # apportate al dataset principale, ocosì da mantenere la coerenza tra il 
        # dataset modificato e tutte le viste pre-calcolate utilizzate da 
        # altre operazioni (ricerche, visualizzazioni, analisi statistiche).
        self._aggiorna_strutture_interne()
```

```python
class AnalisiModel(DatasetArticoliModel):
    def __init__(self, dataset_articoli: dict):
        """
        Classe Model specializzata per l'analisi statistica, descrittiva e metrica del dataset articoli.
        
        Estende 'DatasetArticoliModel' fornendo implementazioni concrete per analisi quantitative
        avanzate che trasformano i dati grezzi in insight significativi su:
        - Conteggi di base (quanti articoli ci sono in totale),
        - Ordine e classifiche (quali articoli sono più o meno visibili),
        - Parole ricorrenti nei titoli (analisi delle parole più usate),
        - Confronti tra categorie e tipi di narrativa (quali si posizionano meglio),
        - Presenza nelle prime posizioni (quali compaiono nelle prime 10 posizioni).
        
        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
                
        Note:
            - Il Model si occupa esclusivamente della logica di calcolo statistico,
              delegando la formattazione e visualizzazione ai metodi ereditati dalla classe base.
            - Tutte le analisi operano sulle strutture dati pre-calcolate della classe base
              per garantire efficienza e consistenza dei risultati.
            - Include un set configurato di stopwords per filtrare termini non significativi
              nelle analisi testuali, migliorando la qualità degli insight lessicali.
        """
        super().__init__(dataset_articoli)

        # set[str] - Insieme di parole comuni da escludere dall'analisi delle frequenze nei titoli.
        # Include articoli determinativi, preposizioni, congiunzioni, pronomi, verbi ausiliari
        # e termini ridondanti specifici del dominio (es. "titolo", "articolo").
        # Utilizzato nel metodo '_parole_piu_frequenti_nei_titoli' per filtrare termini non significativi
        # e focalizzare l'analisi sui contenuti tematici veramente rilevanti.  
        self._stop_words = {
            "di", "a", "da", "in", "con", "su", "per", "tra", "fra",
            "il", "lo", "la", "i", "gli", "le", "un", "uno", "una",
            "e", "o", "ma", "se", "come", "che", "quando",
            "nei", "sui", "alle", "nella", "nelle", "delle",
            "degli", "dai", "agli", "col", "coi",
            "del", "della", "dello", "dei",
            "al", "allo", "ai", "lo",
            "chi", "cosa",
            "d", "l", "un'",
            "sono", "è", "ha", "hanno", "sta", "stanno",
            "si", "ci", "vi", "mi", "ti",
            "dico", "dice",
            "titolo", "articolo", "media" # termini ridondanti.
        }
        
    # ======================== METODI DI ANALISI STATISTICA ========================
    def _conteggio_articoli_totali(self) -> None:
        """
        Calcola e mostra il numero totale di articoli presenti nel dataset.
        
        Fornisce il numero totale di articoli nel dataset. 
        È un dato di base ma importante per capire quanto è grande il dataset 
        e per interpretare correttamente tutte le altre statistiche.
        
        Note:
            - Mostra il risultato direttamente nella console con formattazione chiara.              
            - Utilizza il seguente attributo della classe base:
              * '_articoli_per_pos_crescente' (list[tuple[int, str, str, str]]): lista
                pre-ordinata degli articoli dalla quale viene estratta la lunghezza
                come rappresentazione accurata del conteggio totale.
        """
        console.print(f"Gli articoli presenti nel dataset sono {len(self._articoli_per_pos_crescente)}.")
          
    def _articoli_con_pos_top_bottom_five(self) -> None:
        """
        Trova e mostra gli articoli nelle posizioni più estreme: i 5 più visibili e i 5 meno visibili.
        
        Secondo il principio che numeri più bassi significano maggiore visibilità 
        (1 è la posizione migliore), questo metodo:
        - Seleziona i 5 articoli con posizionamento più basso (più visibili),
        - Seleziona i 5 articoli con posizionamento più alto (meno visibili),
        - Presenta entrambi i gruppi in tabelle separate per confronto immediato.
        
        Note:
            - Si assume che l'ordine crescente per posizione corrisponda alla visibilità: 
              prima posizione = più visibile, ultima posizione = meno visibile.
            - Presenta due tabelle distinte per evitare confusione tra i due gruppi.
            - Include tutte le informazioni contestuali (categoria, narrativa, titolo)
              per permettere analisi qualitative oltre che quantitative.
              
            - Utilizza '_stampa_tabella' (metodo): helper ereditato dalla classe base
              per la creazione e visualizzazione di tabelle formattate.
              
            - Utilizza i seguenti attributi della classe base:
              * '_articoli_per_pos_crescente' (list[tuple[int, str, str, str]]): lista
                pre-ordinata per posizionamento crescente che permette l'estrazione
                efficiente degli estremi tramite slicing.
              * '_COLONNA_POS', '_COLONNA_CATEGORIA', '_COLONNA_NARRATIVA',
                '_COLONNA_TITOLO' (str): costanti per le etichette delle colonne.
              * '_COLONNA_*' (str) per le etichette delle colonne.
        """        
        # Estrae i primi 5 articoli (posizionamento più basso, più visibili)
        # e gli ultimi 5 articoli (posizionamento più alto, meno visibili) tramite slicing.
        # list[tuple[int, str, str, str]] - Formato: (posizione, categoria, narrativa, titolo).   
        articoli_pos_top_five = self._articoli_per_pos_crescente[:5]    # Primi 5: più visibili
        articoli_pos_bottom_five = self._articoli_per_pos_crescente[-5:] # Ultimi 5: meno visibili
        
        # Prepara i dati per la visualizzazione tabellare degli articoli più visibili.
        # list[tuple[int, str, str, str]] - Formato righe: (posizione, categoria, narrativa, titolo)
        righe_top = [
            [pos, categoria.capitalize(), narrativa, titolo]
            for pos, categoria, narrativa, titolo in articoli_pos_top_five
        ]
        
        # Prepara i dati per la visualizzazione tabellare degli articoli meno visibili.
        # list[tuple[int, str, str, str]] - Formato righe: (posizione, categoria, narrativa, titolo)
        righe_bottom = [
            [pos, categoria.capitalize(), narrativa, titolo]
            for pos, categoria, narrativa, titolo in articoli_pos_bottom_five
        ]
        
        # Configurazione colonne comune per entrambe le tabelle, ottimizzata per confronto.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_POS, "center", False),     # Posizionamento centrato, testo a capo automatico.
            (self._COLONNA_CATEGORIA, "right", True), # Categoria allineata a destra, testo su singola riga.
            (self._COLONNA_NARRATIVA, "right", True), # Narrativa allineata a destra, testo su singola riga.
            (self._COLONNA_TITOLO, "left", True)      # Titolo allineato a sinistra, testo su singola riga.
        ]
        
        # Visualizza le due tabelle separate per confronto chiaro degli estremi.
        self._stampa_tabella("I 5 articoli più visualizzati.", colonne, righe_top)
        self._stampa_tabella("I 5 articoli meno visualizzati.", colonne, righe_bottom)
    
    def _parole_piu_frequenti_nei_titoli(self, frequenza_min=10) -> None:
        """
        Analizza e visualizza le parole che compaiono più frequentemente nei titoli.
        
        Implementa un'analisi testuale avanzata che:
        1. Estrae tutte le parole da tutti i titoli degli articoli,
        2. Pulisce il testo (rimozione punteggiatura, lowercase),
        3. Filtra le parole non significative (stopwords, parole corte <4 caratteri),
        4. Calcola le frequenze assolute delle parole rimanenti,
        5. Filtra ulteriormente per soglia di frequenza minima configurabile,
        6. Ordina le parole per frequenza decrescente,
        7. Presenta i risultati in formato tabellare.
        
        Questa analisi rivela i temi dominanti, le parole chiave ricorrenti e
        il vocabolario caratteristico del dataset, aiutando a migliorare i contenuti 
        e a capire quali argomenti sono più trattati.
        
        Args:
            frequenza_min (int): numero minimo di volte che una parola deve comparire 
                                 per essere inclusa nei risultati. Valore predefinito: 10. 
                                 Esclude le parole rare e mostra solo quelle che appaiono frequentemente.
            
        Note:
            - Utilizza il modulo 'string' della libreria standard Python per la
              gestione della punteggiatura (string.punctuation).
            - Utilizza '_stop_words' (set[str]): insieme configurato di parole comuni
              da escludere automaticamente dall'analisi per migliorarne la qualità.
              
            - Utilizza i seguenti metodi della classe base:
              * '_stampa_tabella' (metodo): per la visualizzazione formattata dei risultati.
              * '_get_key_estrazione_valore' (metodo): crea automaticamente le funzioni 
              necessarie per ordinare i risultati dalla frequenza più alta alla più bassa.
                
            - Utilizza '_articoli_per_pos_crescente' (list[tuple[int, str, str, str]]):
              lista pre-ordinata della classe base come sorgente dati per l'analisi testuale.
        """
        # dict{parola (str): conteggio (int)} - Dizionario per accumulare le occorrenze
        # delle parole significative attraverso tutti i titoli.
        conteggio_parole = {} 

        # Analizza iterativamente ogni titolo per estrarre e contare le parole significative.
        for _, _, _, titolo in self._articoli_per_pos_crescente:
            # Fase di pulizia del testo: rimuove tutta la punteggiatura standard
            # sostituendola con spazi, e converte tutto in lowercase per normalizzazione.
            titolo = ''.join(' ' if c in string.punctuation else c for c in titolo).lower()
            parole = titolo.split() # list[str] - Divide il titolo in singole parole, separandole dove ci sono spazi
            
            for parola in parole:
                # Filtraggio delle parole: include solo parole con lunghezza >= 4 caratteri
                # e che non sono nell'insieme delle stopwords configurate.
                if len(parola) >= 4 and parola not in self._stop_words:
                    # Incrementa il contatore per questa parola, inizializzandolo a 0
                    # se la parola non è ancora presente nel dizionario.
                    conteggio_parole[parola] = conteggio_parole.get(parola, 0) + 1

        # Fase di filtraggio post-analisi: seleziona solo le parole che raggiungono
        # la soglia minima di frequenza specificata dal parametro.
        # list[tuple[str, int]] - Formato: (parola, frequenza).
        parole_filtrate = [(parola, count) for parola, count in conteggio_parole.items() if count >= frequenza_min]
        
        # Ordinamento delle parole filtrate per frequenza decrescente per evidenziare
        # i termini più ricorrenti in prima posizione.
        # list[tuple[str, int]] - Formato: (parola, frequenza).
        parole_filtrate_decrescenti = sorted(parole_filtrate, key=self._get_key_estrazione_valore(1), reverse=True)
        
        # Prepara i dati per la visualizzazione tabellare dei risultati.
        # list[list[str]] - Formato righe: (parola, frequenza).
        righe = [
            [parola, frequenza]
            for parola, frequenza in parole_filtrate_decrescenti
        ]
        
        # Configurazione colonne ottimizzata per analisi lessicale.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            ("Parola", "center", False),   # Parola centrata, testo a capo automatico.
            ("Frequenza", "center", True)  # Frequenza centrata, testo su singola riga.
        ]
        
        self._stampa_tabella("Le parole più frequenti nei titoli.", colonne, righe)
    
    def _distribuzione_percentuale_narrativa(self) -> None:
        """
        "Mostra la percentuale di articoli per ogni tipo di narrativa.
        
        Calcola e analizza come sono distribuiti i diversi generi nel dataset.
        I generi sono ordinati dal più frequente al meno frequente, 
        per evidenziare quelli principali e quelli secondari.
        
        Note:
            - Calcola percentuali precise arrotondate a 2 decimali per bilanciare
              precisione e leggibilità.
            - Ordina i risultati per percentuale decrescente per evidenziare
              immediatamente i generi più rappresentati.
            - Mostra sia i valori percentuali che le etichette dei generi per
              interpretazione contestuale completa.
              
            - Utilizza i seguenti metodi della classe base:
              * '_stampa_tabella' (metodo): per la visualizzazione formattata della distribuzione.
              * '_get_key_estrazione_valore' (metodo): per l'ordinamento decrescente per percentuale.
              
            - Utilizza i seguenti attributi della classe base:
              * '_articoli_per_pos_crescente' (list[tuple[int, str, str, str]]): lista
                pre-ordinata come sorgente dati per il conteggio delle occorrenze.
              * '_COLONNA_NARRATIVA' (str): costante per l'etichetta della colonna narrativa.
        """
        # dict{narrativa (str): conteggio (int)} - Dizionario per accumulare le occorrenze
        # di ciascun tipo di narrativa attraverso tutti gli articoli.
        conteggio_narrative = {} 
        
        # Conta iterativamente le occorrenze di ogni tipo di narrativa nel dataset.
        for articolo in self._articoli_per_pos_crescente:
            # Estrae il tipo di narrativa dall'indice 2 della tupla articolo.
            narrativa = articolo[2]
            
            # Gestisce sia il caso in cui la narrativa è già presente nel dizionario
            # (incrementa il conteggio) che il caso in cui è nuova (inizializza a 1).
            if narrativa in conteggio_narrative:
                conteggio_narrative[narrativa] += 1
            else:
                conteggio_narrative[narrativa] = 1
                
        # Forma abbreviata equivalente (commentata per riferimento):
        # for _, _, narrativa, _ in self._articoli_per_pos_crescente:
        #   conteggio_narrative[narrativa] = conteggio_narrative.get(narrativa, 0) + 1
        
        # Calcola la percentuale di presenza per ogni tipo di narrativa
        # rispetto al totale degli articoli nel dataset.
        
        # int - Conteggio totale degli articoli, base per il calcolo percentuale.
        totale_articoli = len(self._articoli_per_pos_crescente)
        
        # list[tuple[str, float]] - Calcola per ogni narrativa: (narrativa, percentuale).
        # La percentuale è arrotondata a 2 decimali per leggibilità.
        distribuzione = [
                (narrativa, round((conteggio / totale_articoli) * 100, 2)) # 2 cifre decimali.
            for narrativa, conteggio in conteggio_narrative.items()
        ]
        
        # Ordina la distribuzione per percentuale decrescente per evidenziare
        # i generi narrativi più rappresentati in prima posizione.
        # list[tuple[str, float]] - Formato: (narrativa, percentuale).
        distribuzione_decrescente = sorted(distribuzione, key=self._get_key_estrazione_valore(1), reverse=True)
        
        # Prepara i dati per la visualizzazione tabellare con formattazione
        # della percentuale (aggiunta del simbolo '%').
        # list[list[str]] - Formato righe: (narrativa, percentuale formattata).
        righe = []
        for narrativa, percentuale in distribuzione_decrescente:
            righe.append([narrativa, f"{percentuale}%"]) 
        
        # Configurazione colonne ottimizzata per visualizzazione di distribuzioni percentuali.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_NARRATIVA, "right", False), # Narrativa allineata a destra, testo a capo automatico.
            ("Frequenza %", "center", False)           # Frequenza percentuale centrata, testo a capo automatico.
        ]
        
        self._stampa_tabella(f"Frequenza percentuale\ndei vari tipi di narrativa.", colonne, righe)
     
    def _pos_medio_per_categoria(self) -> None:
        """
        Mostra la posizione media degli articoli di ogni categoria.
        
        Calcola per ogni categoria la media delle posizioni dei suoi articoli.
        Le categorie sono messe in ordine: prima quelle con posizione media più bassa
        (quindi articoli più in vista), poi quelle con posizione media più alta.
        
        Note:
            - Le medie sono calcolate su tutti gli articoli di ogni categoria,
              fornendo un quadro complessivo delle performance di ogni categoria.
            - L'ordinamento crescente per media posizionamento evidenzia
              immediatamente le categorie con migliore visibilità complessiva.
            - I valori sono arrotondati a 2 decimali per precisione e leggibilità.
              
            - Utilizza:
              * '_calcola_pos_medio_per_etichetta' (metodo): helper specifico 
              che si occupa di calcolare le medie per ogni etichetta.
              * '_stampa_tabella' (metodo): helper ereditato dalla classe base
                per la visualizzazione formattata dei risultati.
                
            - Utilizza '_COLONNA_CATEGORIA' (str): costante della classe base
              per l'etichetta della colonna categoria.
        """
        # Calcola le medie dei posizionamenti per ogni categoria utilizzando
        # il metodo helper specializzato con indice 1 (categoria).
        # list[tuple[str, float]] - Formato: (categoria, media_posizionamento).
        pos_medio_per_categoria = self._calcola_pos_medio_per_etichetta(1)
        
        # Prepara i dati per la visualizzazione tabellare con capitalizzazione
        # delle categorie per uniformità stilistica.
        # list[list[str]] - Formato righe: (categoria capitalizzata, media posizionamento).
        righe = [
            [categoria.capitalize(), media_pos]
            for categoria, media_pos in pos_medio_per_categoria
        ]
        
        # Configurazione colonne ottimizzata per metriche di performance media.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_CATEGORIA, "right", False), # Categoria allineata a destra, testo a capo automatico.
            ("Pos.\nmedio", "left", True)              # Posizionamento medio allineato a sinistra, testo su singola riga.
        ]
        
        self._stampa_tabella("Posizionamento medio per categoria.", colonne, righe)
           
    def _pos_medio_per_narrativa(self) -> None:
        """
        Mostra la posizione media degli articoli di ogni narrativa.
        
        Calcola per ogni narrativa la media delle posizioni dei suoi articoli.
        Le narrative sono messe in ordine: prima quelle con posizione media più bassa
        (quindi articoli più in vista), poi quelle con posizione media più alta.
        
        Note:
            - Le medie sono calcolate su tutti gli articoli di ogni tipo di narrativa,
              fornendo un quadro complessivo delle performance di ogni narrativa.
            - L'ordinamento crescente per media posizionamento evidenzia
              immediatamente i generi narrativi con migliore visibilità complessiva.
            - I valori sono arrotondati a 2 decimali per precisione e leggibilità.
              
            - Utilizza:
              * '_calcola_pos_medio_per_etichetta' (metodo): helper specifico 
              che si occupa di calcolare le medie per ogni etichetta.
              * '_stampa_tabella' (metodo): helper ereditato dalla classe base
                per la visualizzazione formattata dei risultati.
                
            - Utilizza '_COLONNA_NARRATIVA' (str): costante della classe base
              per l'etichetta della colonna narrativa.
        """
        # Calcola le medie dei posizionamenti per ogni tipo di narrativa utilizzando
        # il metodo helper specializzato con indice 2 (tipo di narrativa).
        # list[tuple[str, float]] - Formato: (narrativa, media_posizionamento).
        pos_medio_per_narrativa = self._calcola_pos_medio_per_etichetta(2)
        
        # Prepara i dati per la visualizzazione tabellare mantenendo il formato
        # originale delle narrative (senza capitalizzazione automatica).
        # list[list[str]] - Formato righe: (narrativa, media posizionamento).
        righe = [
            [narrativa, media_pos]
            for narrativa, media_pos in pos_medio_per_narrativa
        ]
        
        # Configurazione colonne ottimizzata per metriche di performance media.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_NARRATIVA, "right", False), # Narrativa allineata a destra, testo a capo automatico.
            ("Pos.\nmedio", "left", True)              # Posizionamento medio allineato a sinistra, testo su singola riga.
        ]

        self._stampa_tabella("Posizionamento medio\nper tipo di narrativa.", colonne, righe)
    
    def _conteggio_categorie_top_ten_pos(self) -> None:
        """
        Mostra quali categorie hanno articoli nelle prime 10 posizioni e quanti per ogni categoria.
        
        Conta gli articoli di ogni categoria che si trovano nelle posizioni 1-10 (le più visibili).
        Le categorie sono ordinate dalla più presente alla meno presente nelle prime posizioni.
        
        Note:
            - Considera esclusivamente i primi 10 posizionamenti (1-10 inclusi) 
              come le posizioni più visibili.
            - Mostra solo le categorie che hanno almeno un articolo nella top ten,
              filtrando automaticamente le categorie assenti da questa fascia.
            - L'ordinamento dal conteggio più alto al più basso fa vedere 
              subito quali categorie compaiono più spesso nelle prime posizioni.
              
            - Utilizza:
              * '_conteggio_etichetta_top_ten_pos' (metodo): helper specializzato
                che si occupa di contare quante volte ogni etichetta compare nelle prime 10 posizioni.
              * '_stampa_tabella' (metodo): helper ereditato dalla classe base
                per la visualizzazione formattata dei risultati.
                
            - Utilizza '_COLONNA_CATEGORIA' (str): costante della classe base
              per l'etichetta della colonna categoria.
        """ 
        # Analizza la presenza delle categorie nei primi 10 posizionamenti utilizzando
        # il metodo helper specializzato con indice 1 (categoria).
        # list[tuple[str, int]] - Formato: (categoria, conteggio articoli in top 10).
        conteggio_categorie_top_ten_pos = self._conteggio_etichetta_top_ten_pos(indice_etichetta=1) 
       
        # Prepara i dati per la visualizzazione tabellare con capitalizzazione
        # delle categorie per uniformità stilistica.
        # list[list[str]] - Formato righe: (categoria capitalizzata, conteggio).
        righe = [
            [categoria.capitalize(), conteggio]
            for categoria, conteggio in conteggio_categorie_top_ten_pos
        ]
        
        # Configurazione colonne ottimizzata per analisi di presenza in posizioni premium.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_CATEGORIA, "right", False), # Categoria allineata a destra, testo a capo automatico.
            ("Conteggio", "center", True)              # Conteggio centrato, testo su singola riga.
        ]
        
        self._stampa_tabella("Conteggio delle categorie presenti\nnei primi 10 posizionamenti.", colonne, righe)
      
    def _conteggio_narrative_top_ten_pos(self) -> None:
        """
        Mostra quali narrative hanno articoli nelle prime 10 posizioni e quanti per ogni narrativa.
        
        Conta gli articoli di ogni narrativa che si trovano nelle posizioni 1-10 (le più visibili).
        Le narrative sono ordinate dalla più presente alla meno presente nelle prime posizioni.
        
        Note:
            - Considera esclusivamente i primi 10 posizionamenti (1-10 inclusi) 
              come le posizioni più visibili.
            - Mostra solo le narrative che hanno almeno un articolo nella top ten,
              filtrando automaticamente le narrative assenti da questa fascia.
            - L'ordinamento dal conteggio più alto al più basso fa vedere 
              subito quali narrative compaiono più spesso nelle prime posizioni.
              
            - Utilizza:
              * '_conteggio_etichetta_top_ten_pos' (metodo): helper specializzato
                che si occupa di contare quante volte ogni etichetta compare nelle prime 10 posizioni.
              * '_stampa_tabella' (metodo): helper ereditato dalla classe base
                per la visualizzazione formattata dei risultati.
                
            - Utilizza '_COLONNA_NARRATIVA' (str): costante della classe base
              per l'etichetta della colonna narrativa.
        """
        # Analizza la presenza dei tipi di narrativa nei primi 10 posizionamenti utilizzando
        # il metodo helper specializzato con indice 2 (tipo di narrativa).
        # list[tuple[str, int]] - Formato: (narrativa, conteggio articoli in top 10).
        conteggio_narrative_top_ten_pos = self._conteggio_etichetta_top_ten_pos(indice_etichetta=2)     
        
        # Prepara i dati per la visualizzazione tabellare mantenendo il formato
        # originale delle narrative (senza capitalizzazione automatica).
        # list[list[str]] - Formato righe: (narrativa, conteggio).
        righe = [
            [narrativa, conteggio]
            for narrativa, conteggio in conteggio_narrative_top_ten_pos
        ]
        
        # Configurazione colonne ottimizzata per analisi di presenza in posizioni premium.
        # list[tuple[str, str, bool]] - Formato: (nome_colonna, allineamento, no_wrap).
        colonne = [
            (self._COLONNA_NARRATIVA, "right", False), # Narrativa allineata a destra, testo a capo automatico.
            ("Conteggio", "center", True)              # Conteggio centrato, testo su singola riga.
        ]

        self._stampa_tabella("Conteggio\ndei tipi di narrativa presenti\nnei primi 10 posizionamenti.", colonne, righe)
    # ======================== METODI HELPER STATISTICI ========================   
    def _calcola_pos_medio_per_etichetta(self, indice_etichetta: int) -> list[tuple[str, float]]:
        """
        Calcola la posizione media per categorie o tipi di narrativa.
        
        Calcola la media delle posizioni degli articoli per ogni categoria o narrativa.
        Una media più bassa significa che gli articoli di quel gruppo tendono a essere più in alto.
        
        Metodo utilizzato sia per le categorie che per le narrative per evitare di ripetere lo stesso codice.

        Args:
            indice_etichetta (int): indica se calcolare per categorie o narrative
                                    1 = categoria, 2 = tipo di narrativa.

        Returns:
            list[tuple[str, float]]: lista di coppie (etichetta, media_pos) ordinata per media crescente.
                                     Le medie più basse (migliori performance) vengono prima.
            
        Note:
            - Utilizza:
              * '_get_key_estrazione_valore' della classe base per ordinare i risultati.
              * '_articoli_per_pos_crescente' della classe base per i dati.
        """
        # Dizionario per raccogliere tutte le posizioni di ogni etichetta.
        # dict{etichetta (str) : posizioni (list[int])}
        pos_per_etichetta = {} 

        # Raccoglie tutte le posizioni degli articoli per ogni etichetta.
        for articolo in self._articoli_per_pos_crescente:
            pos = articolo[0]
            etichetta = articolo[indice_etichetta] # (1=categoria, 2=tipo narrativa)
            # Aggiunge la posizione alla lista di quella etichetta.
            # Se l'etichetta non esiste, crea una lista vuota e poi aggiunge la posizione.
            pos_per_etichetta.setdefault(etichetta, []).append(pos)

        # Calcola la media per ogni etichetta: somma delle posizioni diviso numero di articoli.
        # Arrotonda a 2 decimali per una lettura più facile.
        # list[tuple[str, float]] - (etichetta, media_pos).
        media_per_etichetta = [
            (etichetta, round(sum(pos) / len(pos), 2))
            for etichetta, pos in pos_per_etichetta.items()
        ]
        
        # Ordina per media crescente: medie più basse (articoli più in alto) vengono prima.
        return sorted(media_per_etichetta, key=self._get_key_estrazione_valore(1)) 
    
    def _conteggio_etichetta_top_ten_pos(self, indice_etichetta: int, soglia_pos=10) -> list[tuple[str, int]]:
        """
        Conta quante volte categorie o narrative compaiono nelle prime posizioni.
        
        Metodo che analizza la presenza di categorie o narrative nelle posizioni più visibili.
        Utile per vedere quali gruppi dominano le prime posizioni del dataset.

        Args:
            - indice_etichetta (int): indica se contare per categorie o narrative
                                      1 = categoria, 2 = tipo di narrativa.
            - soglia_pos (int): posizione massima da considerare (predefinito: 10).

        Returns:
            list[tuple[str, int]]: lista di coppie (etichetta, conteggio) ordinata per conteggio decrescente.
                                   Le etichette con più occorrenze nelle prime posizioni vengono prima.
        
        Note:
            - Utilizza:
              * '_get_key_estrazione_valore' della classe base per ordinare i risultati.
              * '_articoli_per_pos_crescente' della classe base per i dati.
        """
        # Dizionario per contare quante volte ogni etichetta compare nelle prime posizioni.
        # dict{etichetta (str) : conteggio (int)}
        conteggio_per_etichetta = {}

        # Controlla ogni articolo nelle posizioni entro il limite specificato.
        for articolo in self._articoli_per_pos_crescente:
            pos = articolo[0]
            etichetta = articolo[indice_etichetta] # (1=categoria, 2=tipo narrativa)
            # Considera solo gli articoli nelle prime posizioni (entro la soglia).
            if pos <= soglia_pos:
                # Aumenta il conteggio per questa etichetta.
                # Se l'etichetta non esiste, parte da 0 e poi aggiunge 1.
                conteggio_per_etichetta[etichetta] = conteggio_per_etichetta.get(etichetta, 0) + 1

        # Ordina per conteggio decrescente per mostrare prima le etichette più frequenti.
        return sorted(conteggio_per_etichetta.items(), key=self._get_key_estrazione_valore(1), reverse=True)
```

### Classi View

#### Introduzione.

In questa sezione sono raccolte le classi che costituiscono l’interfaccia utente del sistema:  
gestiscono i menù, le istruzioni e i feedback testuali per l’interazione con l’utente.  

Tutte le classi View condividono una struttura comune che consente di separare le responsabilità nel progetto.  
La classe madre `DatasetArticoliView` fornisce i menù principali e le funzionalità di base per la visualizzazione delle  opzioni,  
mentre le classi figlie estendono questa logica per introdurre interfacce specifiche per ogni tipo di operazione:  

- `RicercaUnitariaView`: gestisce l’interfaccia per la ricerca di articoli basata su un singolo criterio.  
- `RicercaCombinataView`: gestisce l’interfaccia per la ricerca avanzata combinando più criteri.  
- `VisualizzazioneView`: gestisce la presentazione del dataset completo in base a criteri selezionati.  
- `RegistrazioneView`: permette l’interfaccia guidata per l’inserimento di nuovi articoli.  
- `AnalisiView`: gestisce la visualizzazione di dati statistici e riepiloghi numerici sul dataset.  

Questo gruppo di classi prende il nome "View" perché rappresenta l’interfaccia utente del sistema,  
distinta dalla logica applicativa (Model) e dalla gestione del flusso di controllo (Controller). 

#### Classe `DatasetArticoliView` - base.

```python
class DatasetArticoliView:
    def __init__(self, dataset_articoli: dict):
        """
        View base per la gestione dell'interfaccia testuale utente del sistema.
        
        Contiene la logica comune per la visualizzazione di tutti i menu,
        le istruzioni e i messaggi di feedback dell'applicazione.
        È progettata per essere estesa da classi figlie specializzate.

        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
        """
        # dict - struttura dati principale contenente il dataset completo degli articoli.
        self._dataset_articoli = dataset_articoli
        
        # dict - interfaccia del menu principale, punto di ingresso dell'applicazione.
        self._menu_principale = {
            '00': f'Esci dal programma.'.upper() + f'{USCITA_PROGRAMMA}',
            1: f'Ricerca articoli per criterio singolo.' + f'{RICERCA_SINGOLA}',
            2: f'Ricerca articoli per criteri multipli.'f'{RICERCA_MULTIPLA}',
            3: f'Visualizza il dataset completo di articoli.' + f'{VISUALIZZA_DATASET}',
            4: f'Registra un nuovo articolo.' + f'{NUOVO_ELEMENTO}',
            5: f'Visualizza dati statistici.' + f'{DATI_STATISTICI}'
        }
        
        # dict - categorie disponibili nel dataset con opzioni di navigazione integrate.
        self._elenco_categorie_e_scelte_extra = {
            -1: f'Torna al menu precedente.'.upper() + f'{TORNA_INDIETRO}',
            0: f'Torna alla schermata iniziale.'.upper() + f'{HOME_MENU}',
            '00': f'Esci dal programma.'.upper() + f'{USCITA_PROGRAMMA}', 
            1: 'Attivismo e diritti delle donne',
            2: 'Cultura e genere',
            3: 'Disparità sociali',
            4: 'Donne nel mondo',
            5: 'Educazione alla parità',
            6: 'Intersezionalità',
            7: 'Lavoro e disparità salariali',
            8: 'Parità di genere in Italia',
            9: 'Rappresentazione nei media',
            10: 'Stereotipi e ruoli di genere',
            11: 'Salute e genere'
        }
    
        # dict - tipi di narrativa disponibili nel dataset con opzioni di navigazione integrate.
        self._elenco_narrative_e_scelte_extra = {
            -1: f'Torna al menu precedente.'.upper() + f'{TORNA_INDIETRO}',
            0: f'Torna alla schermata iniziale.'.upper() + f'{HOME_MENU}',
            '00': f'Esci dal programma.'.upper() + f'{USCITA_PROGRAMMA}',
            1: 'Analisi',
            2: 'Approfondimento',
            3: 'Attivismo',
            4: 'Inchiesta',
            5: 'Reportage',
            6: 'Studio',
            7: 'Testimonianza',
        }
        
        # str - istruzioni standard per la navigazione tra i menu, comune a tutti i contesti.
        self._istruzione_scelte_extra = (
            f"Oppure digita:\n"
            f"'-1' per tornare al menu precedente. '{TORNA_INDIETRO}'\n"
            f"'0' per tornare alla schermata iniziale. '{HOME_MENU}'\n"
            f"'00' per uscire dal programma. '{USCITA_PROGRAMMA}'"
        )

        # str - prompt standardizzato per l'acquisizione dell'input utente.
        self._input_prompt = '>> Segui le istruzioni sul terminale e inserisci la tua scelta: '
           
        # dict - messaggi di feedback standardizzati per interazioni utente.
        self._feedback_ui = {
            'errore': f'{ERRORE} Scelta non valida. Riprova. {ERRORE}',
            'uscita': f"{USCITA_PROGRAMMA} Programma interrotto dall'utente. {USCITA_PROGRAMMA}",
            'torna_indietro': f'{TORNA_INDIETRO} Torna al menu precedente. {TORNA_INDIETRO}',
            'schermata_iniziale': f'{HOME_MENU} Torna alla schermata iniziale. {HOME_MENU}'
        }
        
    # ======================== PROPRIETÀ DINAMICHE ========================
    @property
    def _pos_min(self) -> int:
        """
        Calcola e restituisce il posizionamento minimo presente nel dataset.
            
        Returns: 
            int: posizionamento minimo trovato tra tutti gli articoli.
            
        Note:
            - Utilizza:
              * '_dataset_articoli' (dict): struttura dati principale contenente il dataset.
        """
        return min(
            articolo['Posizionamento']
            for categoria, articoli in self._dataset_articoli.items()
            for articolo in articoli
        )

    @property
    def _pos_max(self) -> int:
        """
        Calcola e restituisce il posizionamento massimo presente nel dataset.
                    
        Returns: 
            int: posizionamento massimo trovato tra tutti gli articoli.
        
        Note:
            - Utilizza:
              * '_dataset_articoli' (dict): struttura dati principale contenente il dataset.
        """
        return max(
            articolo['Posizionamento']
            for categoria, articoli in self._dataset_articoli.items()
            for articolo in articoli
        )     
           
    # ======================== METODI ========================
    def _mostra_menu_principale(self) -> None:
        """
        Visualizza il menu principale dell'applicazione.
        
        Note:
            - Utilizza:
              * '_menu_principale' (dict): configurazione opzioni menu principale.
              * '_stampa_menu' (callable): metodo helper per formattazione e visualizzazione.
        """
        self._stampa_menu(self._menu_principale, "Seleziona l'operazione desiderata:")
        
    def _mostra_operazione_scelta(self, operazione: str) -> None:
        """
        Visualizza l'operazione selezionata dall'utente per conferma contestuale.

        Args: 
            operazione: descrizione testuale dell'operazione scelta.
        """
        console.print(f"{INPUT_UTENTE} Operazione selezionata:",  f"'{operazione}'")        
       
    def _mostra_menu_ricerca_per_categoria(self) -> None:
        """
        Visualizza il menu per la selezione della categoria di articoli.
        
        Note:
            - Utilizza:
              * '_elenco_categorie_e_scelte_extra' (dict): configurazione categorie disponibili.
              * '_stampa_menu' (callable): metodo helper per formattazione e visualizzazione.
        """
        self._stampa_menu(self._elenco_categorie_e_scelte_extra, "Seleziona la categoria desiderata:")   
        
    def _mostra_categoria_scelta(self, categoria: str) -> None:
        """
        Visualizza la categoria selezionata dall'utente per conferma contestuale.

        Args: 
            categoria: nome della categoria scelta dall'utente.
        """
        console.print(f"{INPUT_UTENTE} Categoria scelta dall'utente:", f"'{categoria.capitalize()}'.")
        
    def _mostra_menu_ricerca_per_narrativa(self) -> None:
        """
        Visualizza il menu per la selezione del tipo di narrativa.
        
        Note:
            - Utilizza:
              * '_elenco_narrative_e_scelte_extra' (dict): configurazione tipi narrativa disponibili.
              * '_stampa_menu' (callable): metodo helper per formattazione e visualizzazione.
        """
        self._stampa_menu(self._elenco_narrative_e_scelte_extra, "Seleziona il tipo di narrativa desiderata:")
        
    def _mostra_narrativa_scelta(self, narrativa: str) -> None:
        """
        Visualizza il tipo di narrativa selezionato dall'utente per conferma contestuale.

        Args: 
            narrativa: tipo di narrativa scelto dall'utente.
        """
        console.print(f"{INPUT_UTENTE} Narrativa scelta dall'utente:", f"'{narrativa.capitalize()}'.")  
    
    # ======================== METODI HELPER ========================
    def _stampa_menu(self, dizionario_menu: dict, titolo: str) -> None:
        """
        Formatta e visualizza un menu con titolo e opzioni numerate.

        Args: 
            - dizionario_menu: dizionario contenente le opzioni del menu.
            - titolo: titolo descrittivo da mostrare sopra il menu.
        """
        # Formatta il titolo in maiuscolo per enfasi visiva.
        titolo_testuale = titolo.upper() 
        
        # Applica formattazione avanzata al titolo per migliorare la leggibilità.
        titolo_formattato = f"{ATTENZIONE} [bold cyan]{titolo_testuale}[/bold cyan] {ATTENZIONE}\n"
        
        # Crea una linea separatrice proporzionata alla lunghezza del titolo.
        linea_separatrice = "[cyan]" + '-' * (len(titolo_testuale) + 7) + "[/cyan]\n"
        
        # Costruisce la struttura completa del menu.
        menu = titolo_formattato + linea_separatrice
        
        # Itera attraverso le opzioni del menu, formattando ciascuna con numero e descrizione.
        for numero, descrizione in dizionario_menu.items():
            menu += f"{numero}. {descrizione}\n"
        
        console.print(menu)
```

#### Classe `RicercaUnitariaView` - derivata.

### Classi Controller

#### Introduzione.

In questa sezione sono raccolte le classi che costituiscono il cuore del flusso di controllo dell'applicazione:  
gestiscono la navigazione tra i menu, coordinano l'interazione tra Model e View, e implementano la logica di stato del programma.

Tutte le classi Controller condividono una struttura gerarchica che consente di separare le responsabilità nel progetto.  
La classe madre `DatasetArticoliController` fornisce le funzionalità di base per la gestione del dataset,    
la validazione degli input e la navigazione tra i menu, mentre le classi figlie estendono questa logica per gestire operazioni specifiche:  

- `RicercaUnitariaController`: coordina il flusso di ricerche basate su un singolo criterio.  
- `RicercaCombinataController`: gestisce ricerche avanzate che combinano più criteri contemporaneamente.  
- `VisualizzazioneController`: controlla la presentazione ordinata del dataset completo.  
- `RegistrazioneController`: supervisiona il processo guidato di inserimento nuovi articoli.  
- `AnalisiController`: dirige la visualizzazione di metriche statistiche e riepiloghi analitici.  
- `HomeController`: funge da punto di ingresso unico, coordinando tutti i controller specializzati.  

Questo gruppo di classi implementa il pattern "State Transfer Object" attraverso classi segnale specializzate:  

- `ReturnToNearestMenu`: per tornare al menu gerarchicamente più vicino.  
- `GoToMainMenu`: per ritornare al menu principale dell'applicazione.  
- `ExitProgram`: per terminare l'applicazione in modo controllato.  

Questo gruppo di classi prende il nome "Controller" perché rappresenta la logica di controllo del sistema,  
distinta dalla gestione dei dati (Model) e dall'interfaccia utente (View), implementando efficacemente il pattern MVC.

```python
class DatasetArticoliController:
    def __init__(self, dataset_articoli: dict, funzione_input=input):
        """
        Controller base per la gestione della logica applicativa e del flusso di controllo.
        
        Coordina le operazioni sul dataset, gestisce la navigazione tra i menu
        e media l'interazione tra Model e View secondo il pattern MVC.
        
        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
        """
        # dict - Dataset principale su cui operano tutti i metodi del controller.
        self._dataset_articoli = dataset_articoli
        
        # callable - Attributo d'istanza che contiene la funzione input standard, configurabile per testing, di default è 'input'.
        self._input = funzione_input
        
        # str - Codici speciali per la navigazione tra i menu.
        self._scelta_torna_indietro ='-1'
        self._scelta_schermata_iniziale = '0'
        self._scelta_uscita = '00'
        
        # list[str] - Lista unificata di tutte le scelte di navigazione speciali.
        self._SCELTE_SPECIALI = [self._scelta_torna_indietro, self._scelta_schermata_iniziale, self._scelta_uscita]
        
        # ATTRIBUTO DI COMPOSIZIONE con la View per la gestione dell'interfaccia utente.
        self._view = DatasetArticoliView(dataset_articoli)
              
    # ======================== METODI ========================
    def _elabora_input_menu(
        self, 
        mostra_interfaccia: callable, 
        menu: dict, 
        scelte_speciali: list[str] = None
    ) -> str | ReturnToNearestMenu | GoToMainMenu | ExitProgram:
        """
        Gestisce il ciclo completo di interazione con un menu dell'applicazione.
        
        Visualizza l'interfaccia, acquisisce l'input utente e valida la scelta
        secondo le opzioni disponibili e le scelte speciali di navigazione.

        Args:
            - mostra_interfaccia (callable): metodo callable per visualizzare il menu specifico.
            - menu (dict): dizionario contenente le opzioni numerate del menu.
            - scelte_speciali list[str]: lista personalizzabile di scelte di navigazione speciali
                                         (default: tutte e tre le scelte standard).

        Returns:
            Stringa con la scelta valida oppure istanza di:
            - ReturnToNearestMenu per tornare al menu precedente,
            - GoToMainMenu per tornare al menu principale,
            - ExitProgram per terminare l'applicazione.
            
        Note:
            - Utilizza:
              * '_scelta_torna_indietro', '_scelta_schermata_iniziale', '_scelta_uscita' (str) scelte extra selezionabili dall'utente.
              * '_validazione_scelta_menu' metodo che gestisce la validazione e elaborazione dell'input.
        """
        # Utilizza le scelte speciali standard se non ne vengono specificate di personalizzate.
        if scelte_speciali is None:
            scelte_speciali = self._SCELTE_SPECIALI
            
        # Visualizza l'interfaccia del menu tramite il metodo fornito.
        mostra_interfaccia() 
        
        # Breve pausa per migliorare l'esperienza utente e garantire
        # una visualizzazione fluida tra menu e messaggi di feedback.
        time.sleep(0.05)
        
        # Delegata la validazione e gestione dell'input al metodo specializzato.
        return self._validazione_scelta_menu(menu, scelte_speciali) 
             
    def _validazione_scelta_menu(
        self, 
        menu: dict, 
        scelte_speciali: list[str]
    ) -> str | ReturnToNearestMenu | GoToMainMenu | ExitProgram:
        """
        Verifica la validità dell'input utente rispetto alle opzioni di menu e scelte speciali.
        
        Implementa il ciclo di acquisizione e validazione input, gestendo
        sia le scelte ordinarie che i comandi di navigazione speciali.

        Args:
            - menu (dict): dizionario contenente le opzioni numerate del menu.
            - scelte_speciali (list[str]): lista delle scelte di navigazione speciali accettate.

        Returns:
            Scelta validata come stringa o comando di navigazione speciale.

        Note:
            - Utilizza:
              * '_input': funzione per l'acquisizione input utente.
              * '_view._input_prompt': prompt standardizzato per l'input.
              * '_view._feedback_ui['errore']': messaggio di errore per input non valido.
              
              * '_gestisci_scelta_speciale': metodo che processa i comandi di navigazione speciali.
        """
        # list[str] - Estrae le chiavi numeriche del menu come stringhe per il confronto.
        opzioni_menu = [str(numero) for numero in menu.keys()]

        while True:
            # Acquisisce input utente con prompt standardizzato.
            # '>> Segui le istruzioni sul terminale e inserisci la tua scelta: '.
            scelta = self._input(self._view._input_prompt).strip() 
            
            # Gestisce scelte speciali di navigazione.
            if scelta in scelte_speciali:
                return self._gestisci_scelta_speciale(scelta)
            # Gestisce scelte ordinarie del menu.
            elif scelta in opzioni_menu:
                return scelta
            # Gestisce input non validi.
            else:
                # 'Scelta non valida. Riprova.' - e richiede nuovamente l'input.
                console.print(self._view._feedback_ui['errore'])
                # Breve pausa per migliorare l'esperienza utente e garantire
                # una visualizzazione fluida tra menu e messaggi di feedback.
                time.sleep(0.05)
            
    def _gestisci_scelta_speciale(self, scelta: str) -> ReturnToNearestMenu | GoToMainMenu | ExitProgram:
        """
        Trasforma una scelta speciale nel corrispondente comando di navigazione.
        
        Args:
            - scelta (str): codice speciale inserito dall'utente ('-1', '0', o '00').

        Returns:
            Istanza del comando di navigazione corrispondente.

        Note:
            - Utilizza:
              * '_view._feedback_ui': dizionario contenente i messaggi di feedback per ogni tipo di scelta speciale.'
        """
        if scelta == '-1':
            console.print(self._view._feedback_ui['torna_indietro']) # 'Torna al menù precedente.'
            return ReturnToNearestMenu()
        elif scelta == '0':
            console.print(self._view._feedback_ui['schermata_iniziale']) # ' Torna alla schermata iniziale.'
            return GoToMainMenu()
        elif scelta == '00':
            console.print(self._view._feedback_ui['uscita']) # 'Programma interrotto dall'utente.'
            return ExitProgram()  
        
    def _gestisci_due_scelte_speciali(self, scelta: str) -> GoToMainMenu | ExitProgram | None:
        """
        Filtra e restituisce solo i comandi di navigazione di livello superiore.
        
        Utilizzato in contesti dove la navigazione non consente di tornare indietro al menu precedente,
        ma solo di ritornare al menu principale o uscire dall'applicazione.

        Args:
            - scelta (str): input da verificare.

        Returns:
            Comando di navigazione di livello superiore o None se non applicabile.
        """
        if isinstance(scelta, (GoToMainMenu, ExitProgram)):
            return scelta
        return None
       
    def _gestisci_tre_scelte_speciali(self, scelta: str) -> ReturnToNearestMenu | GoToMainMenu | ExitProgram | None:
        """
        Filtra e restituisce qualsiasi comando di navigazione speciale.
    
        Utilizzato in contesti dove la navigazione consente sia di tornare indietro al menu precedente,
        che di ritornare al menu principale o uscire dall'applicazione.
        
        Args:
            - scelta (str): input da verificare.
            
        Returns:
            Comando di navigazione corrispondente o None se non speciale.
        """
        if isinstance(scelta, (ReturnToNearestMenu, GoToMainMenu, ExitProgram)):
            return scelta
        return None
          
    def _estrai_categoria_da_scelta(self, scelta: str) -> str:
        """
        Converte una scelta numerica nel nome della categoria corrispondente.
        
        Args:
            - scelta (str): numero di selezione della categoria.

        Returns:
            Nome della categoria in formato uppercase.

        Note:
            - Utilizza:
              * '_view._elenco_categorie_e_scelte_extra': dizionario di mappatura numeri → nomi categorie
        """
        # Converte l’indice numerico nella categoria corrispondente.
        num = int(scelta)
        categoria = self._view._elenco_categorie_e_scelte_extra[num].upper()
        return categoria
    
    def _estrai_narrativa_da_scelta(self, scelta: str) -> str:
        """
        Converte una scelta numerica nel tipo di narrativa corrispondente.
        
        Args:
            - scelta (str): numero di selezione della narrativa.

        Returns:
            Nome del tipo di narrativa.

        Note:
            - Utilizza:
              * '_view._elenco_narrative_e_scelte_extra': dizionario di mappatura numeri → tipi di narrativa.
        """
        # Converte l’indice numerico nella narrativa corrispondente.
        num = int(scelta)
        narrativa = self._view._elenco_narrative_e_scelte_extra[num]
        return narrativa
    
    def _rileva_presenza_parola_generico(
        self,
        parola: str,
        funzione_rilevazione: str,
        filtro: str = None
    ) -> bool:
        """
        Coordinatore generico per le ricerche di parole nei titoli degli articoli di una categoria o di una narrativa specifica.
        
        Args:
            - parola (str): termine da cercare nei titoli degli articoli
            - funzione_rilevazione (str): identificativo della strategia di ricerca da utilizzare.
            - filtro (str): parametro opzionale per delimitare il contesto di ricerca (categoria o tipo di narrativa).

        Returns:
            True se la parola viene trovata in almeno un titolo, False altrimenti.

        Note:
            - La ricerca è case-insensitive.
            - Il parametro 'funzione_rilevazione' deve corrispondere a una delle strategie predefinite supportate.
            
            - Utilizza i metodi:
              * '_parola_presente_nei_titoli': ricerca generale in tutti i titoli o filtrata per tipo di narrativa.
              * '_parola_presente_in_categoria': ricerca filtrata per categoria.
        """
        # Normalizza la parola per ricerca case-insensitive.
        parola = parola.lower().strip()

        # Delega la richiesta al metodo helper appropriato.
        if funzione_rilevazione == "nei_titoli":
            return self._parola_presente_nei_titoli(parola)
        elif funzione_rilevazione == "nei_titoli_per_narrativa":
            return self._parola_presente_nei_titoli(parola, filtro)
        elif funzione_rilevazione == "in_categoria":
            return self._parola_presente_in_categoria(parola, filtro)
    
    # ======================== METODI HELPER ========================
    def _parola_presente_nei_titoli(self, parola: str, narrativa: str = None) -> bool:
        """
        Verifica la presenza di una parola nei titoli degli articoli.
        
        Esegue una ricerca testuale completa su tutti i titoli del dataset,
        con opzione di filtraggio per tipo di narrativa specifico.

        Args:
            - parola (str): termine da cercare (ricerca case-insensitive).
            - narrativa (str): tipo di narrativa per delimitare la ricerca (opzionale).

        Returns:
            True se la parola compare in almeno un titolo che soddisfa i criteri,
            False se non viene trovata in nessun titolo.

        Attr:
            _view._dataset_articoli: Dataset completo degli articoli su cui effettuare la ricerca

        Note:
            - Se narrativa è None, la ricerca include tutti gli articoli.
            - L'implementazione utilizza una comprehension per efficienza con dataset di medie dimensioni.
        """
        return any(
            parola in articolo['Titolo'].lower()
            for articoli in self._view._dataset_articoli.values()
            for articolo in articoli
            if narrativa is None or articolo['TipoNarrativa'] == narrativa
        )
        
    def _parola_presente_in_categoria(self, parola: str, categoria_scelta: str) -> bool:
        """
        Verifica la presenza di una parola nei titoli degli articoli di una specifica categoria.
        
        Esegue una ricerca testuale sui titoli degli articoli appartenenti a una categoria specifica.

        Args:
            - parola (str): termine da cercare (ricerca case-insensitive).
            - categoria (str): nome della categoria in cui delimitare la ricerca.

        Returns:
            True se la parola compare in almeno un titolo della categoria specificata,
            False se non viene trovata o se la categoria non esiste.

        Note:
            - La ricerca è case-insensitive.
            - Utilizza:
              * '_view._dataset_articoli': dataset completo degli articoli su cui effettuare la ricerca.
        """
        # Cerca la parola nei titoli degli articoli della categoria specificata.
        return any(
            parola in articolo['Titolo'].lower()
            for categoria, articoli in self._view._dataset_articoli.items()
            if categoria_scelta == categoria
            for articolo in articoli
        )
```

#### Classe `RicercaUnitariaController` - derivata.

```python
class RicercaUnitariaController(DatasetArticoliController):
    def __init__(self, dataset_articoli: dict, funzione_input=input):
        """   
        Controller specializzato per le ricerche di articoli basate su singolo criterio.
        
        Gestisce il flusso di interazione per ricerche per categoria, posizionamento,
        tipo di narrativa e parole nei titoli, coordinando View e Model specializzati.    
        
        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
        """
        super().__init__(dataset_articoli, funzione_input)
        
        # ATTRIBUTI DI COMPOSIZIONE con View e Model specializzati per ricerche unitarie.
        self._view = RicercaUnitariaView(dataset_articoli) # Interfaccia testuale per ricerche singole.
        self._model = RicercaUnitariaModel(dataset_articoli) # Logica operativa per ricerche singole.
        
        # str - descrizione dell'operazione corrente per il contesto utente.
        # f'Ricerca articoli per criterio singolo.' + f'{RICERCA_SINGOLA}'
        self._operazione = self._view._menu_principale[1] 

    # ======================== METODI ========================
    def _ricerca_criterio_singolo(self) -> GoToMainMenu | ExitProgram:
        """
        Gestisce il flusso principale delle ricerche per criterio singolo.
        
        Coordina la navigazione tra le diverse modalità di ricerca unitaria
        e gestisce i comandi di navigazione di livello superiore.

        Returns:
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_scelta_schermata_iniziale' (str): codice per tornare al menu principale.
                  * '_scelta_uscita' (str): codice per terminare l'applicazione.
                  * '_elabora_input_menu' (callable): metodo di gestione acquisizione e validazione input utente.
                  * '_gestisci_due_scelte_speciali' (callable): metodo di filtro comandi di navigazione superiore.
                
                - Di composizione:
                * '_view._mostra_menu_ricerca_singolo_criterio' (callable): visualizzazione menu ricerche.
                * '_view._menu_ricerca_singolo_criterio' (dict): configurazione opzioni ricerca.
        """
        # dict - Mappatura tra selezione utente e metodi di ricerca specializzati.
        # L'utente può selezionare il criterio per effettuare la ricerca unitaria.
        azioni = {
            # Metodi per accedere e interagire con menù specifici:
            '1': self._ricerca_per_categoria, # per la ricerca di articoli in base alla categoria.
            '2': self._ricerca_per_pos_singolo, # per la ricerca di un articolo in base al posizionamento.
            '3': self._ricerca_per_narrativa, # per la ricerca di articoli in base al tipo di narrativa.
            '4': self._ricerca_per_parola_nei_titoli, # per la ricerca di articoli in base a una parola nei titoli.
        }

        while True:
            # Mostra il contesto operativo corrente all'utente.
            # f'Ricerca articoli per criterio singolo.' + f'{RICERCA_SINGOLA}'
            self._view._mostra_operazione_scelta(self._operazione)
            
            # Gestisce l'interazione con il menu delle ricerche unitarie.
            scelta = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_ricerca_singolo_criterio,
                menu=self._view._menu_ricerca_singolo_criterio,
                scelte_speciali=[self._scelta_schermata_iniziale, self._scelta_uscita]
            )

            # Verifica e gestisce comandi di navigazione di livello superiore.
            scelta_speciale = self._gestisci_due_scelte_speciali(scelta)
            if scelta_speciale is not None:
                return scelta_speciale
            
            # Esegue la ricerca specializzata corrispondente alla selezione utente.
            # 'risultato' corrisponderà a None se il metodo ha terminato normalmente, con nessun codice di stato.
            risultato = azioni[scelta]()
                        
            # Gestisce il risultato della ricerca (comandi di navigazione)
            if isinstance(risultato, ReturnToNearestMenu):
                continue # Ritorna al menu ricerche unitarie.
            elif isinstance(risultato, GoToMainMenu):
                return GoToMainMenu() # Ritorna al menu principale.
            elif isinstance(risultato, ExitProgram):
                return ExitProgram() # Termina l'applicazione.

    def _ricerca_per_categoria(self) -> None | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Gestisce il flusso di ricerca articoli per categoria.
        
        Coordina la selezione della categoria e l'esecuzione della ricerca
        tramite il Model specializzato.

        Returns:
            - None per ricerca completata normalmente.
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.
            - ReturnToNearestMenu per ritornare al menu precedente.

        Utilizza:
            - Della classe base:
              * '_elabora_input_menu' (callable): gestione input selezione categoria.
              * '_gestisci_tre_scelte_speciali' (callable): gestione comandi navigazione completa.
              * '_estrai_categoria_da_scelta' (callable): conversione selezione → nome categoria.
            
            - Di composizione:
              * '_view._mostra_menu_ricerca_per_categoria'(callable): visualizzazione menu categorie.
              * '_view._elenco_categorie_e_scelte_extra' (dict): configurazione categorie disponibili.
              * '_view._menu_ricerca_singolo_criterio' (dict): descrizione criterio corrente.
              * '_view._mostra_criterio_scelto' (callable): feedback criterio selezionato.
              
              * '_model._ricerca_per_categoria' (callable): esecuzione ricerca effettiva.
        """
        while True:
            # Mostra il criterio di ricerca corrente.
            # 'Ricerca per categoria.'
            criterio = self._view._menu_ricerca_singolo_criterio[1]
            self._view._mostra_criterio_scelto(criterio)
            
            # Gestisce selezione categoria da parte dell'utente.
            scelta = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_ricerca_per_categoria,
                menu=self._view._elenco_categorie_e_scelte_extra
                # Essendo stata lasciata a None contiene tutte e tre le scelte speciali disponibili. 
                # scelte_speciali=[self.scelta_torna_indietro, self.scelta_schermata_iniziale, self.scelta_uscita]
            )

            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_tre_scelte_speciali(scelta)
            if scelta_speciale is not None:
                return scelta_speciale
            
            # Converte la selezione in nome categoria ed esegue la ricerca.
            categoria = self._estrai_categoria_da_scelta(scelta)

            self._model._ricerca_per_categoria(categoria)
            
    def _ricerca_per_pos_singolo(self) -> None | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Gestisce il flusso di ricerca articoli per posizionamento singolo.
        
        Guida l'utente nell'inserimento del posizionamento e coordina la ricerca
        dell'articolo corrispondente tramite il Model specializzato.

        Returns:
            - None se la ricerca è stata completata con successo.
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.
            - ReturnToNearestMenu per ritornare al menu precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_input' (callable): funzione per gestire l'input dell'utente, di default è 'input'.
                  * '_gestisci_scelta_speciale' (callable): metodo di gestione comandi di navigazione speciali.
                  
                - Di composizione:
                  * '_view._menu_ricerca_singolo_criterio' (dict): operazione specifica selezionata dall'utente.
                  * '_view._input_prompt' (str): prompt di input comune a tutti i menu e le istruzioni.
                  * '_view._feedback_ui' (dict): messaggio di errore per scelta non valida.
                  * '_view._mostra_criterio_scelto' (callable): feedback criterio selezionato.
                  * '_view._mostra_istruzione_ricerca_per_pos_singolo' (callable): guida per l'inserimento posizionamento.
                  
                  * '_model._pos_articoli' (list[int]): lista dei posizionamenti disponibili nel dataset.
                  * '_model._ricerca_per_pos_singolo' (callable): esecuzione ricerca per posizionamento.
        """
        # bool - variabile di controllo per mostrare l'istruzione solo all'inizio e dopo input valido.
        # Evita di mostrare l'istruzione subito dopo un input errato, ma la ripresenta dopo una ricerca completata.
        mostra_istruzione = True
        
        while True:
            # 'Ricerca per posizionamento singolo.'
            criterio = self._view._menu_ricerca_singolo_criterio[2]
            
            if mostra_istruzione:
                # Mostra il criterio di ricerca corrente per tenere traccia del contesto.
                self._view._mostra_criterio_scelto(criterio)
                # Presenta all'utente le istruzioni per l'inserimento del posizionamento.
                self._view._mostra_istruzione_ricerca_per_pos_singolo()
                
                # Breve pausa per migliorare l'esperienza utente e garantire
                # una visualizzazione fluida tra menu e messaggi di feedback.
                time.sleep(0.05)
                mostra_istruzione = False
                
            # Acquisisce l'input dell'utente tramite il prompt standardizzato.
            # '>> Segui le istruzioni sul terminale e inserisci la tua scelta: '
            scelta = self._input(self._view._input_prompt).strip()

            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_scelta_speciale(scelta)
            if scelta_speciale is not None:
                return scelta_speciale
            
            # Se la scelta corrisponde a un posizionamento valido nel dataset,
            # converte in intero ed esegue la ricerca tramite il Model.
            if scelta in str(self._model._pos_articoli):
                pos = int(scelta)
                self._model._ricerca_per_pos_singolo(pos)
                mostra_istruzione = True  # Dopo l'input valido, ristampa l’istruzione. 
            else:
                # Mostra messaggio di errore per input non valido.
                # f'{ERRORE} Scelta non valida. Riprova. {ERRORE}'
                console.print(self._view._feedback_ui['errore'])
                # Breve pausa per migliorare l'esperienza utente e garantire
                # una visualizzazione fluida tra menu e messaggi di feedback.
                time.sleep(0.05)

    def _ricerca_per_narrativa(self) -> None | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Gestisce il flusso di ricerca articoli per tipo di narrativa.
        
        Coordina la selezione del tipo di narrativa e l'esecuzione della ricerca
        tramite il Model specializzato.

        Returns:
            - None se la ricerca è stata completata con successo.
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.
            - ReturnToNearestMenu per ritornare al menu precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_elabora_input_menu' (callable): gestione input selezione narrativa.
                  * '_gestisci_tre_scelte_speciali' (callable): gestione comandi navigazione completa.
                  * '_estrai_narrativa_da_scelta' (callable): conversione selezione → tipo narrativa.
                  
                - Di composizione:
                  * '_view._mostra_menu_ricerca_per_narrativa' (callable): visualizzazione menu narrative.
                  * '_view._elenco_narrative_e_scelte_extra' (dict): configurazione tipi narrativa disponibili.
                  * '_view._menu_ricerca_singolo_criterio' (dict): descrizione criterio corrente.
                  * '_view._mostra_criterio_scelto' (callable): feedback criterio selezionato.
                  
                  * '_model._ricerca_per_narrativa' (callable): esecuzione ricerca per tipo narrativa.
        """
        while True:
            # Mostra il criterio di ricerca corrente per tenere traccia del contesto.
            # 'Ricerca per tipo di narrativa.'
            criterio = self._view._menu_ricerca_singolo_criterio[3]
            self._view._mostra_criterio_scelto(criterio)
            
            # Gestisce selezione tipo di narrativa da parte dell'utente.
            # Utilizza tutte e tre le scelte speciali disponibili per default.
            scelta = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_ricerca_per_narrativa,
                menu=self._view._elenco_narrative_e_scelte_extra
                #scelte_speciali=[self.scelta_torna_indietro, self.scelta_schermata_iniziale, self.scelta_uscita]
            )

            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_tre_scelte_speciali(scelta)
            if scelta_speciale is not None:
                return scelta_speciale

            # Converte la selezione in tipo di narrativa ed esegue la ricerca.
            narrativa = self._estrai_narrativa_da_scelta(scelta)

            self._model._ricerca_per_narrativa(narrativa)
            
    def _ricerca_per_parola_nei_titoli(self) -> None | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Gestisce il flusso di ricerca articoli per parola nei titoli.
        
        Guida l'utente nell'inserimento della parola da cercare e coordina la ricerca
        degli articoli corrispondenti tramite il Model specializzato.

        Returns:
            - None se la ricerca è stata completata con successo.
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.
            - ReturnToNearestMenu per ritornare al menu precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_input' (callable): funzione per gestire l'input dell'utente, di default è 'input'.
                  * '_gestisci_scelta_speciale' (callable): gestione comandi di navigazione speciali.
                  * '_rileva_presenza_parola_generico' (callable): verifica presenza parola nei titoli.
                  
                - Di composizione:
                  * '_view._menu_ricerca_singolo_criterio' (dict): operazione specifica selezionata dall'utente.
                  * '_view._input_prompt' (str): prompt di input comune a tutti i menu e le istruzioni.
                  * '_view._feedback_ui' (dict): messaggio di errore per scelta non valida.
                  * '_view._mostra_criterio_scelto' (callable): feedback criterio selezionato.
                  * '_view._mostra_istruzione_ricerca_per_parola_nei_titoli' (callable): guida per l'inserimento parola.
                  * '_view._mostra_feedback_errore_parola' (callable): messaggio errore parola non trovata.
                  
                  * '_model._ricerca_per_parola_nei_titoli' (callable): esecuzione ricerca per parola nei titoli.
        """
        # bool - variabile di controllo per mostrare l'istruzione solo all'inizio e dopo input valido.
        # Evita di mostrare l'istruzione subito dopo un input errato, ma la ripresenta dopo una ricerca completata.
        mostra_istruzione = True
        
        while True:
            # 'Ricerca per parola nei titoli.'
            criterio = self._view._menu_ricerca_singolo_criterio[4]
            
            if mostra_istruzione:
                # Mostra il criterio di ricerca corrente per tenere traccia del contesto.
                self._view._mostra_criterio_scelto(criterio)
                # Presenta all'utente le istruzioni per l'inserimento della parola da cercare.
                self._view._mostra_istruzione_ricerca_per_parola_nei_titoli()
                # Breve pausa per migliorare l'esperienza utente e garantire
                # una visualizzazione fluida tra menu e messaggi di feedback.
                time.sleep(0.05)  
                mostra_istruzione = False
                
            # Acquisisce l'input dell'utente tramite il prompt standardizzato.
            # '>> Segui le istruzioni sul terminale e inserisci la tua scelta: '
            scelta = self._input(self._view._input_prompt).strip()

            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_scelta_speciale(scelta)
            if scelta_speciale is not None:
                return scelta_speciale
            
            # Verifica se la parola inserita è presente in almeno un titolo del dataset.
            # bool - indica se la parola è stata trovata nei titoli.
            parola_trovata = self._rileva_presenza_parola_generico(
                parola=scelta,
                funzione_rilevazione="nei_titoli"
            )

            if parola_trovata:
                parola = scelta
                self._model._ricerca_per_parola_nei_titoli(parola)
                mostra_istruzione = True  # Dopo input valido, ristampa l’istruzione.
            else:
                # Mostra messaggio di errore per parola non trovata nei titoli.
                # 'La parola inserita non compare in nessun titolo. Riprova.'
                self._view._mostra_feedback_errore_parola()
                # Breve pausa per migliorare l'esperienza utente e garantire
                # una visualizzazione fluida tra menu e messaggi di feedback.
                time.sleep(0.05)               
```

```python
class RicercaCombinataController(DatasetArticoliController):
    def __init__(self, dataset_articoli: dict, funzione_input=input):
        """
        Controller specializzato per le ricerche di articoli basate su criteri multipli.
        
        Gestisce il flusso di interazione per ricerche combinate che includono range di posizionamenti,
        categorie con narrative specifiche e parole nei titoli con filtri aggiuntivi,
        coordinando View e Model specializzati.
        
        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }  
        """
        super().__init__(dataset_articoli, funzione_input)
        
        # ATTRIBUTI DI COMPOSIZIONE con View e Model specializzati per ricerche combinate.
        self._view = RicercaCombinataView(dataset_articoli) # Interfaccia testuale per ricerche combinate.
        self._model = RicercaCombinataModel(dataset_articoli) # Logica operativa per ricerche combinate.
        
        # str - descrizione dell'operazione corrente per il contesto utente.
        # f'Ricerca articoli per criteri multipli.' + f'{RICERCA_MULTIPLA}'
        self._operazione = self._view._menu_principale[2] 
        
    # ======================== METODI ========================
    def _ricerca_criteri_multipli(self) -> GoToMainMenu | ExitProgram:
        """
        Gestisce il flusso principale delle ricerche per criteri multipli.
        
        Coordina la navigazione tra le diverse modalità di ricerca combinata
        e gestisce i comandi di navigazione di livello superiore.

        Returns:
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_scelta_schermata_iniziale' (str): codice per tornare al menu principale.
                  * '_scelta_uscita' (str): codice per terminare l'applicazione.
                  * '_elabora_input_menu' (callable): metodo di gestione acquisizione e validazione input utente.
                  * '_gestisci_due_scelte_speciali' (callable): metodo di filtro comandi di navigazione superiore.
                  
                - Di composizione:
                  * '_view._mostra_menu_ricerca_criteri_multipli' (callable): visualizzazione menu ricerche combinate.
                  * '_view._menu_ricerca_criteri_multipli' (dict): configurazione opzioni ricerca combinata.
                  * '_view._mostra_operazione_scelta' (callable): feedback contesto operazione corrente.
        """
        # dict - Mappatura tra selezione utente e metodi di ricerca specializzati.
        # L'utente può selezionare i criteri per effettuare la ricerca combinata.
        azioni = {
            # Metodi per accedere e interagire con menu specifici:
            '1': self._ricerca_per_range_pos, # per la ricerca di articoli in base al range di posizionamenti.
            '2': self._ricerca_per_categoria_e_narrativa, # per la ricerca di articoli in base a categoria e narrativa correlata.
            '3': self._ricerca_per_categoria_e_parola_nei_titoli, # per la ricerca di articoli in base a categoria e parola nei titoli correlata.
            '4': self._ricerca_per_narrativa_e_parola_nei_titoli # per la ricerca di articoli in base a narrativa e parola nei titoli correlata.
        }

        while True:
            # Mostra il contesto operativo corrente all'utente.
            # f'Ricerca articoli per criteri multipli.' + f'{RICERCA_MULTIPLA}'
            self._view._mostra_operazione_scelta(self._operazione)
            
            # Gestisce l'interazione con il menu delle ricerche combinate.
            scelta = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_ricerca_criteri_multipli,
                menu=self._view._menu_ricerca_criteri_multipli,
                scelte_speciali=[self._scelta_schermata_iniziale, self._scelta_uscita]
            )

            # Verifica e gestisce comandi di navigazione di livello superiore.
            scelta_speciale = self._gestisci_due_scelte_speciali(scelta)
            if scelta_speciale is not None:
                return scelta_speciale
            
            # Esegue la ricerca specializzata corrispondente alla selezione utente.
            # 'risultato' corrisponderà a None se il metodo ha terminato normalmente, con nessun codice di stato.
            risultato = azioni[scelta]()
            
            # Gestisce il risultato della ricerca (comandi di navigazione).
            if isinstance(risultato, ReturnToNearestMenu):
                continue # Ritorna al menu ricerche combinate.
            elif isinstance(risultato, GoToMainMenu):
                return GoToMainMenu() # Ritorna al menu principale.
            elif isinstance(risultato, ExitProgram):
                return ExitProgram() # Termina l'applicazione.
      
    def _ricerca_per_range_pos(self) -> None | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Gestisce il flusso di ricerca articoli per range di posizionamenti.
        
        Guida l'utente nell'inserimento sequenziale del posizionamento minimo e massimo,
        e coordina la ricerca degli articoli nel range specificato tramite il Model specializzato.

        Returns:
            - None se la ricerca è stata completata con successo.
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.
            - ReturnToNearestMenu per ritornare al menu precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_rileva_pos_valido' (callable): metodo di validazione input posizionamento.
                  * '_gestisci_tre_scelte_speciali' (callable): gestione comandi navigazione completa.
                  
                - Di composizione:
                  * '_view._menu_ricerca_criteri_multipli' (dict): operazione specifica selezionata dall'utente.
                  * '_view._pos_min' (int): posizionamento minimo disponibile nel dataset.
                  * '_view._pos_max' (int): posizionamento massimo disponibile nel dataset.
                  * '_view._mostra_criteri_scelti' (callable): feedback criteri selezionati.
                  * '_view._mostra_istruzione_ricerca_per_range_pos_min' (callable): guida per inserimento posizionamento minimo.
                  * '_view._mostra_istruzione_ricerca_per_range_pos_max' (callable): guida per inserimento posizionamento massimo.
                  * '_view._mostra_pos_min_scelto' (callable): feedback posizionamento minimo selezionato.
                  
                  * '_model._ricerca_per_range_pos' (callable): esecuzione ricerca per range posizionamenti.
        """
        while True:
            # Mostra il criterio di ricerca corrente per tenere traccia del contesto.
            # 'Ricerca per range di posizionamenti.'
            criteri = self._view._menu_ricerca_criteri_multipli[1]
            self._view._mostra_criteri_scelti(criteri) 
            
            # Gestisce acquisizione e validazione del posizionamento minimo.
            scelta = self._rileva_pos_valido(
                mostra_istruzione=self._view._mostra_istruzione_ricerca_per_range_pos_min(),
                valid_range=range(self._view._pos_min, self._view._pos_max)
            )

            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_tre_scelte_speciali(scelta)
            if scelta_speciale is not None:
                return scelta_speciale

            # --- POSIZIONAMENTO MINIMO ---
            pos_min_utente = scelta
            # Mostra conferma del posizionamento minimo selezionato per il contesto utente.
            # f"Posizionamento minimo scelto: {pos_min}."
            self._view._mostra_pos_min_scelto(pos_min_utente)
            
            while True:
                # Gestisce acquisizione e validazione del posizionamento massimo.
                scelta = self._rileva_pos_valido(
                    mostra_istruzione=self._view._mostra_istruzione_ricerca_per_range_pos_max(),
                    valid_range=range(pos_min_utente + 1, self._view._pos_max + 1)
                )

                # Gestisce comandi di navigazione durante l'inserimento del massimo.
                if isinstance(scelta, ReturnToNearestMenu):
                    break # Ritorna all'inserimento del posizionamento minimo.
                elif isinstance(scelta, GoToMainMenu):
                    return GoToMainMenu() # Ritorna al menu principale.
                elif isinstance(scelta, ExitProgram):
                    return ExitProgram() # Termina l'applicazione.

                # --- POSIZIONAMENTO MASSIMO ---
                pos_max_utente = scelta
                # Esegue la ricerca degli articoli nel range specificato dall'utente.
                self._model._ricerca_per_range_pos(pos_min_utente, pos_max_utente)
                break  # Ritorna all'inserimento del posizionamento minimo per una nuova ricerca.
      
    def _ricerca_per_categoria_e_narrativa(self) -> None | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """ 
        Gestisce il flusso di ricerca articoli per categoria e tipo di narrativa correlato.
        
        Coordina la selezione sequenziale della categoria e del tipo di narrativa disponibile
        per quella specifica categoria, eseguendo la ricerca combinata tramite il Model specializzato.

        Returns:
            - None se la ricerca è stata completata con successo.
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.
            - ReturnToNearestMenu per ritornare al menu precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_elabora_input_menu' (callable): gestione input selezione categoria/narrativa.
                  * '_gestisci_tre_scelte_speciali' (callable): gestione comandi navigazione completa.
                  * '_estrai_categoria_da_scelta' (callable): conversione selezione → nome categoria.
                  
                - Di composizione:
                  * '_view._menu_ricerca_criteri_multipli' (dict): operazione specifica selezionata dall'utente.
                  * '_view._mostra_menu_ricerca_per_categoria' (callable): visualizzazione menu categorie.
                  * '_view._elenco_categorie_e_scelte_extra' (dict): configurazione categorie disponibili.
                  * '_view._mostra_menu_ricerca_per_narrativa_da_categoria' (callable): visualizzazione menu narrative per categoria.
                  * '_view._mostra_criteri_scelti' (callable): feedback criteri selezionati.
                  * '_view._mostra_categoria_scelta' (callable): feedback categoria selezionata.
                  * '_view._get_narrative_per_categoria' (callable): estrazione narrative disponibili per categoria.
                  * '_view._aggiorna_elenco_narrativa_in_categoria' (callable): costruzione menu narrative contestuale.
                  
                  * '_model._ricerca_per_categoria_e_narrativa' (callable): esecuzione ricerca combinata.
        """
        while True:
            # Mostra il criterio di ricerca corrente per tenere traccia del contesto.
            # 'Ricerca per categoria e tipo di narrativa.'
            criteri = self._view._menu_ricerca_criteri_multipli[2]
            self._view._mostra_criteri_scelti(criteri) 
            
            # --- CATEGORIA ---
            # Gestisce selezione categoria da parte dell'utente.
            scelta = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_ricerca_per_categoria,
                menu=self._view._elenco_categorie_e_scelte_extra
                # scelte_speciali=[self.scelta_torna_indietro, self.scelta_schermata_iniziale, self.scelta_uscita]
            )

            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_tre_scelte_speciali(scelta)
            if scelta_speciale is not None:
                return scelta_speciale
            
            # Converte la selezione in nome categoria.
            categoria = self._estrai_categoria_da_scelta(scelta)
                 
            # Estrae i tipi di narrativa disponibili per la categoria selezionata.
            narrative_disponibili = self._view._get_narrative_per_categoria(categoria)
            narrative_disponibili = sorted(narrative_disponibili)
            
            # Costruisce il menu contestuale delle narrative per la categoria specifica.
            elenco_narrative_in_categoria_scelta = self._view._aggiorna_elenco_narrativa_in_categoria(narrative_disponibili)
            
            while True:
                # Mostra conferma della categoria selezionata per il contesto utente.
                # f'Categoria scelta dall'utente: {categoria}'
                self._view._mostra_categoria_scelta(categoria)
                # --- NARRATIVA IN CATEGORIA SELEZIONATA ---
                # Gestisce selezione tipo di narrativa da parte dell'utente.
                scelta = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_ricerca_per_narrativa_da_categoria,
                menu=elenco_narrative_in_categoria_scelta
                )
                
                # Gestisce comandi di navigazione durante la selezione della narrativa.
                if isinstance(scelta, ReturnToNearestMenu):
                    break # Ritorna alla selezione della categoria.
                elif isinstance(scelta, GoToMainMenu):
                    return GoToMainMenu() # Ritorna al menu principale.
                elif isinstance(scelta, ExitProgram):
                    return ExitProgram() # Termina l'applicazione.

                # Estrae il tipo di narrativa corrispondente alla selezione.
                num = int(scelta)
                narrativa = elenco_narrative_in_categoria_scelta[num]
                
                # Esegue la ricerca combinata per categoria e tipo di narrativa specificati.
                self._model._ricerca_per_categoria_e_narrativa(categoria, narrativa)
                
    def _ricerca_per_categoria_e_parola_nei_titoli(self) -> None | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Gestisce il flusso di ricerca articoli per categoria e parola nei titoli.
        
        Coordina la selezione della categoria e l'inserimento della parola da cercare
        nei titoli degli articoli di quella specifica categoria, eseguendo la ricerca
        combinata tramite il Model specializzato.

        Returns:
            - None se la ricerca è stata completata con successo.
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.
            - ReturnToNearestMenu per ritornare al menu precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_elabora_input_menu' (callable): gestione input selezione categoria.
                  * '_gestisci_tre_scelte_speciali' (callable): gestione comandi navigazione completa.
                  * '_estrai_categoria_da_scelta' (callable): conversione selezione → nome categoria.
                  * '_rileva_presenza_parola_generico' (callable): verifica presenza parola in categoria.
                  
                - Di composizione:
                  * '_view._menu_ricerca_criteri_multipli' (dict): operazione specifica selezionata dall'utente.
                  * '_view._mostra_menu_ricerca_per_categoria' (callable): visualizzazione menu categorie.
                  * '_view._elenco_categorie_e_scelte_extra' (dict): configurazione categorie disponibili.
                  * '_view._mostra_criteri_scelti' (callable): feedback criteri selezionati.
                  * '_view._mostra_categoria_scelta' (callable): feedback categoria selezionata.
                  * '_view._mostra_istruzione_ricerca_per_parola_nei_titoli_da_categoria' (callable): guida per inserimento parola.
                  * '_view._mostra_feedback_errore_parola_categoria' (callable): messaggio errore parola non trovata.
                  
                  * '_model._ricerca_per_categoria_e_parola_nei_titoli' (callable): esecuzione ricerca combinata. 
        """
        while True:
            # Mostra il criterio di ricerca corrente per tenere traccia del contesto.
            # 'Ricerca per parola nei titoli di una categoria a scelta.'
            criteri = self._view._menu_ricerca_criteri_multipli[3]
            self._view._mostra_criteri_scelti(criteri) 
            # --- CATEGORIA ---
            # Gestisce selezione categoria da parte dell'utente.
            scelta = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_ricerca_per_categoria,
                menu=self._view._elenco_categorie_e_scelte_extra
            )

            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_tre_scelte_speciali(scelta)
            if scelta_speciale is not None:
                return scelta_speciale

            # Converte la selezione in nome categoria.
            categoria = self._estrai_categoria_da_scelta(scelta)
            
            # bool - controllo per mostrare l'istruzione solo all'inizio e dopo input valido.
            prima_iterazione = True
            
            while True:
                # Mostra conferma della categoria selezionata per il contesto utente.
                self._view._mostra_categoria_scelta(categoria)
                # --- PAROLA IN CATEGORIA SELEZIONATA---
                if prima_iterazione:
                    # Presenta all'utente le istruzioni per l'inserimento della parola.
                    self._view._mostra_istruzione_ricerca_per_parola_nei_titoli_da_categoria()
                    # Breve pausa per garantire una visualizzazione fluida tra istruzione e input.
                    time.sleep(0.05)
                    prima_iterazione = False
                
                # Acquisisce l'input dell'utente tramite il prompt standardizzato.
                # '>> Segui le istruzioni sul terminale e inserisci la tua scelta: '
                scelta = self._input(self._view._input_prompt).strip()
                    
                # Verifica e gestisce comandi di navigazione.
                scelta_speciale = self._gestisci_scelta_speciale(scelta)
                if scelta_speciale is not None:
                    return scelta_speciale

                # Verifica se la parola è presente nei titoli della categoria specificata.
                parola_trovata = self._rileva_presenza_parola_generico(
                    parola=scelta,
                    funzione_rilevazione="in_categoria",
                    filtro=categoria
                )

                if parola_trovata:
                    # Esegue la ricerca combinata per categoria e parola specificati.
                    parola = scelta
                    self._model._ricerca_per_categoria_e_parola_nei_titoli(categoria, parola)  
                    prima_iterazione = True  # Dopo input valido, ristampa l’istruzione.
                else:   
                    # Mostra messaggio di errore per parola non trovata nella categoria.
                    # 'La parola inserita non compare in nessun titolo della categoria selezionata. Riprova.'
                    self._view._mostra_feedback_errore_parola_categoria()
                    # Breve pausa per garantire una visualizzazione fluida tra messaggio e nuovo input.
                    time.sleep(0.05)

    def _ricerca_per_narrativa_e_parola_nei_titoli(self) -> None | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Gestisce il flusso di ricerca articoli per tipo di narrativa e parola nei titoli.
        
        Coordina la selezione del tipo di narrativa e l'inserimento della parola da cercare
        nei titoli degli articoli di quella specifica narrativa, eseguendo la ricerca
        combinata tramite il Model specializzato.

        Returns:
            - None se la ricerca è stata completata con successo.
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.
            - ReturnToNearestMenu per ritornare al menu precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_elabora_input_menu' (callable): gestione input selezione narrativa.
                  * '_gestisci_tre_scelte_speciali' (callable): gestione comandi navigazione.
                  * '_estrai_narrativa_da_scelta' (callable): conversione selezione → tipo narrativa.
                  * '_rileva_presenza_parola_generico' (callable): verifica presenza parola in narrativa.
                  
                - Di composizione:
                  * '_view._menu_ricerca_criteri_multipli' (dict): operazione specifica selezionata dall'utente.
                  * '_view._mostra_menu_ricerca_per_narrativa' (callable): visualizzazione menu narrative.
                  * '_view._elenco_narrative_e_scelte_extra' (dict): configurazione tipi narrativa disponibili.
                  * '_view._mostra_criteri_scelti' (callable): feedback criteri selezionati.
                  * '_view._mostra_narrativa_scelta' (callable): feedback narrativa selezionata.
                  * '_view._mostra_istruzione_ricerca_per_parola_nei_titoli_da_narrativa' (callable): guida per inserimento parola.
                  * '_view._mostra_feedback_errore_parola_narrativa' (callable): messaggio errore parola non trovata.
                  
                  * '_model._ricerca_per_narrativa_e_parola_nei_titoli' (callable): esecuzione ricerca combinata.  
        """
        while True:
            # Mostra il criterio di ricerca corrente per tenere traccia del contesto.
            # 'Ricerca per parola nei titoli di un tipo di narrativa a scelta.'
            criteri = self._view._menu_ricerca_criteri_multipli[4]
            self._view._mostra_criteri_scelti(criteri) 
            # --- NARRATIVA ---
            # Gestisce selezione tipo di narrativa da parte dell'utente.
            scelta = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_ricerca_per_narrativa,
                menu=self._view._elenco_narrative_e_scelte_extra
            )

            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_tre_scelte_speciali(scelta)
            if scelta_speciale is not None:
                return scelta_speciale

            # Converte la selezione in tipo di narrativa.
            narrativa = self._estrai_narrativa_da_scelta(scelta)
               
            # bool - controllo per mostrare l'istruzione solo all'inizio e dopo input valido.
            prima_iterazione = True
            
            while True:
                # Mostra conferma del tipo di narrativa selezionato per il contesto utente.
                self._view._mostra_narrativa_scelta(narrativa)
                
                # --- PAROLA IN NARRATIVA SELEZIONATA---
                if prima_iterazione:
                    # Presenta all'utente le istruzioni per l'inserimento della parola.
                    self._view._mostra_istruzione_ricerca_per_parola_nei_titoli_da_narrativa()
                    
                    # Breve pausa per garantire una visualizzazione fluida tra istruzione e input.
                    time.sleep(0.05) 
                    prima_iterazione = False
                
                # Acquisisce l'input dell'utente tramite il prompt standardizzato.
                # '>> Segui le istruzioni sul terminale e inserisci la tua scelta: '
                scelta = self._input(self._view._input_prompt).strip()
                    
                # Verifica e gestisce comandi di navigazione.
                scelta_speciale = self._gestisci_scelta_speciale(scelta)
                if scelta_speciale is not None:
                    return scelta_speciale

                # Verifica se la parola è presente nei titoli della narrativa specificata.
                parola_trovata = self._rileva_presenza_parola_generico(
                    parola=scelta,
                    funzione_rilevazione="nei_titoli_per_narrativa",
                    filtro=narrativa
                )

                if parola_trovata:
                    # Esegue la ricerca combinata per narrativa e parola specificati.
                    parola = scelta
                    self._model._ricerca_per_narrativa_e_parola_nei_titoli(narrativa, parola)
                    prima_iterazione = True  # Ripristina l'istruzione per eventuali ricerche successive.
                else:   
                    # Mostra messaggio di errore per parola non trovata nella narrativa.
                    self._view._mostra_feedback_errore_parola_narrativa()
                    
                    # Breve pausa per garantire una visualizzazione fluida tra messaggio e nuovo input.
                    time.sleep(0.05)
                    
    # ======================== METODI HELPER ========================
    def _rileva_pos_valido(
        self, 
        mostra_istruzione: callable, 
        valid_range: range
    ) -> int | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Helper per l'acquisizione e validazione di posizionamenti inseriti dall'utente.
        
        Gestisce l'interazione completa per l'inserimento di un posizionamento valido,
        includendo la visualizzazione delle istruzioni, la validazione nel range specificato
        e la gestione dei comandi di navigazione speciali.

        Args:
            - mostra_istruzione (callable): funzione per visualizzare le istruzioni di inserimento.
            - valid_range (range): intervallo di posizionamenti validi accettabili.

        Returns:
            - int: posizionamento valido inserito dall'utente.
            - GoToMainMenu: per ritornare al menu principale.
            - ExitProgram: per terminare l'applicazione.
            - ReturnToNearestMenu: per ritornare al menu precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_input' (callable): funzione per gestire l'input dell'utente, di default è 'input'.
                  * '_gestisci_scelta_speciale' (callable): gestione comandi di navigazione speciali.
                  
                - Di composizione:
                  * '_view._input_prompt' (str): prompt di input comune a tutti i menu e le istruzioni.
                  * '_view._feedback_ui' (dict): messaggio di errore per scelta non valida.
        """
        # Visualizza le istruzioni specifiche per l'inserimento del posizionamento.
        mostra_istruzione
        # Breve pausa per migliorare l'esperienza utente e garantire
        # una visualizzazione fluida tra menu e messaggi di feedback.
        time.sleep(0.05)

        while True:
            # Acquisisce l'input dell'utente tramite il prompt standardizzato.
            # '>> Segui le istruzioni sul terminale e inserisci la tua scelta: '
            scelta = self._input(self._view._input_prompt).strip()

            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_scelta_speciale(scelta)
            if scelta_speciale is not None:
                return scelta_speciale

            # Verifica se l'input è un numero valido all'interno del range specificato.
            if scelta.isdigit() and int(scelta) in valid_range:
                return int(scelta)
            else:
                # Mostra messaggio di errore per input non valido.
                # Non ripete l'istruzione per evitare ridondanza visiva.
                # 'Scelta non valida. Riprova.'
                console.print(self._view._feedback_ui['errore'])
                
                # Breve pausa per migliorare l'esperienza utente e garantire
                # una visualizzazione fluida tra menu e messaggi di feedback.
                time.sleep(0.05)
```

```python
class RegistrazioneController(DatasetArticoliController):
    def __init__(self, dataset_articoli: dict, funzione_input=input):
        """
        Controller specializzato per la registrazione di nuovi articoli nel dataset.
        
        Gestisce il flusso completo di inserimento di nuovi articoli, inclusa la validazione
        del titolo e la gestione delle selezioni di categoria e tipo di narrativa,
        coordinando View e Model specializzati.
        
        Args:
            dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }  
        """
        super().__init__(dataset_articoli, funzione_input)
        
        # ATTRIBUTI DI COMPOSIZIONE con View e Model specializzati per registrazione.
        self._view = RegistrazioneView(dataset_articoli)  # interfaccia testuale per registrazione
        self._model = RegistrazioneModel(dataset_articoli)  # logica operativa per registrazione
        
        # str - descrizione dell'operazione corrente per il contesto utente.
        # 'Registra un nuovo articolo.'
        self._operazione = self._view._menu_principale[4] 

    # ======================== METODI ========================
    def _inserisci_nuovo_articolo(self)  -> GoToMainMenu | ExitProgram | None:
        """
        Gestisce il flusso principale della registrazione di nuovi articoli.
        
        Coordina il ciclo completo di registrazione, inclusa la gestione della
        continuazione dopo un inserimento completato con successo.

        Returns:
            - GoToMainMenu per ritornare al menu principale.
            - ExitProgram per terminare l'applicazione.
            - None per continuare con una nuova registrazione.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_gestisci_tre_scelte_speciali' (callable): gestione comandi navigazione.
                  * '_input' (callable): funzione per gestire l'input dell'utente.
                  
                - Di composizione:
                  * '_view._mostra_operazione_scelta' (callable): feedback contesto operazione corrente.
                  * '_view._input_prompt' (str): prompt di input standardizzato.
                  * '_view._mostra_istruzione_continua_registrazione' (callable): richiesta continuazione registrazione.
                  * '_view._mostra_feedback_registrazione_terminata' (callable): conferma fine registrazione.
                  * '_view._mostra_feedback_errore_continua_registrazione' (callable): messaggio errore input continuazione.
        """
        while True:
            # Mostra il contesto operativo corrente all'utente.
            # 'Registra un nuovo articolo.'
            self._view._mostra_operazione_scelta(self._operazione)
            
            # bool - flag per tracciare il successo della registrazione.
            self._articolo_registrato = False
            
            # Esegue il flusso completo di registrazione dell'articolo.
            risultato = self._flusso_registrazione_articolo() 

            # Verifica e gestisce comandi di navigazione dal flusso di registrazione.
            scelta_speciale = self._gestisci_tre_scelte_speciali(risultato)
            if scelta_speciale is not None:
                return scelta_speciale

            # Gestisce la continuazione della registrazione dopo un inserimento completato.
            if self._articolo_registrato:
                while True:
                    # Richiede all'utente se desidera registrare un altro articolo.
                    self._view._mostra_istruzione_continua_registrazione()
                    scelta = self._input(self._view._input_prompt).strip().upper()
                    if scelta == 'Y':
                        break  # Riparte il ciclo per una nuova registrazione.
                    elif scelta == 'N':
                        # Conferma la fine del processo di registrazione.
                        self._view._mostra_feedback_registrazione_terminata()
                        return  # Ritorna al menu principale.
                    else:
                        # Gestisce input non valido per la continuazione.
                        self._view._mostra_feedback_errore_continua_registrazione()    
            
            else:
                return  # Ritorna al menu principale in caso di anomalie o interruzioni.
    
    # ======================== METODI HELPER ========================            
    def _flusso_registrazione_articolo(self) -> None | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Gestisce il flusso sequenziale di registrazione di un singolo articolo.
        
        Coordina la selezione gerarchica di categoria, tipo di narrativa e titolo,
        con validazione completa di tutti gli input.

        Returns:
            - None per registrazione completata con successo.
            - GoToMainMenu per ritorno al menu principale.
            - ExitProgram per terminazione applicazione.
            - ReturnToNearestMenu per ritorno al passo precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_elabora_input_menu' (callable): gestione input selezione categoria/narrativa.
                  * '_gestisci_tre_scelte_speciali' (callable): gestione comandi navigazione.
                  * '_estrai_categoria_da_scelta' (callable): conversione selezione → nome categoria.
                  * '_estrai_narrativa_da_scelta' (callable): conversione selezione → tipo narrativa.
                  * '_validazione_scelta_titolo' (callable): validazione e normalizzazione titolo.
                  
                - Di composizione:
                  * '_view._mostra_menu_seleziona_categoria_per_nuovo_articolo' (callable): visualizzazione menu categorie.
                  * '_view._elenco_categorie_e_scelte_extra_per_titolo_nuovo' (dict): configurazione categorie disponibili.
                  * '_view._mostra_menu_seleziona_narrativa_per_nuovo_articolo' (callable): visualizzazione menu narrative.
                  * '_view._elenco_narrative_e_scelte_extra' (dict): configurazione tipi narrativa disponibili.
                  * '_view._mostra_categoria_scelta' (callable): feedback categoria selezionata.
                  * '_view._mostra_narrativa_scelta' (callable): feedback narrativa selezionata.
                  * '_view._mostra_istruzione_inserisci_nuovo_titolo_articolo' (callable): guida per inserimento titolo.
                  
                  * '_model._registra_articolo' (callable): esecuzione registrazione articolo.
        """
        while True:
            # --- CATEGORIA ---
            # Gestisce selezione categoria da parte dell'utente.
            scelta_categoria = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_seleziona_categoria_per_nuovo_articolo,
                menu=self._view._elenco_categorie_e_scelte_extra_per_titolo_nuovo,
                scelte_speciali=[self._scelta_schermata_iniziale, self._scelta_uscita]
            )
            
            # Verifica e gestisce comandi di navigazione.
            scelta_speciale = self._gestisci_tre_scelte_speciali(scelta_categoria)
            if scelta_speciale is not None:
                return scelta_speciale

            # Converte la selezione in nome categoria.
            categoria = self._estrai_categoria_da_scelta(scelta_categoria)

            while True:
                # Mostra conferma della categoria selezionata.
                self._view._mostra_categoria_scelta(categoria)
                
                # --- NARRATIVA ---
                # Gestisce selezione tipo di narrativa da parte dell'utente.
                scelta_narrativa = self._elabora_input_menu(
                    mostra_interfaccia=self._view._mostra_menu_seleziona_narrativa_per_nuovo_articolo,
                    menu=self._view._elenco_narrative_e_scelte_extra
                )

                # Gestisce comandi di navigazione durante la selezione della narrativa.
                if isinstance(scelta_narrativa, ReturnToNearestMenu):
                    break # Ritorna alla selezione della categoria.
                elif isinstance(scelta_narrativa, GoToMainMenu):
                    return GoToMainMenu() # Ritorna al menu principale.
                elif isinstance(scelta_narrativa, ExitProgram):
                    return ExitProgram() # Termina l'applicazione.

                # Converte la selezione in tipo di narrativa.
                narrativa = self._estrai_narrativa_da_scelta(scelta_narrativa)

                while True:
                    # Mostra conferma del tipo di narrativa selezionato.
                    self._view._mostra_narrativa_scelta(narrativa)
                    
                    # Presenta all'utente le istruzioni per l'inserimento del titolo.
                    self._view._mostra_istruzione_inserisci_nuovo_titolo_articolo()
                    
                    # Gestisce acquisizione e validazione del titolo.
                    scelta = self._validazione_scelta_titolo()
                    
                    # Gestisce comandi di navigazione durante l'inserimento del titolo.
                    if isinstance(scelta, ReturnToNearestMenu):
                        break # Ritorna alla selezione della narrativa.
                    elif isinstance(scelta, GoToMainMenu):
                        return GoToMainMenu() # Ritorna al menu principale.
                    elif isinstance(scelta, ExitProgram):
                        return ExitProgram() # Termina l'applicazione.

                    # Esegue la registrazione finale dell'articolo.
                    titolo = scelta
                    self._model._registra_articolo(categoria, narrativa, titolo)
                    self._articolo_registrato = True
                    return # Registrazione completata con successo.
                
    def _validazione_scelta_titolo(self) -> str | GoToMainMenu | ExitProgram | ReturnToNearestMenu:
        """
        Helper per la validazione completa del titolo di un nuovo articolo.
        
        Implementa tutti i controlli di validazione per garantire che il titolo
        rispetti le specifiche tecniche e grammaticali richieste.

        Returns:
            - str: titolo validato e normalizzato.
            - GoToMainMenu: per ritorno al menu principale.
            - ExitProgram: per terminazione applicazione.
            - ReturnToNearestMenu: per ritorno al passo precedente.

        Note:
            - Utilizza:
                - Della classe base:
                  * '_input' (callable): funzione per gestire l'input dell'utente.
                  * '_gestisci_scelta_speciale' (callable): gestione comandi navigazione.
                  
                - Di composizione:
                  * '_view._input_prompt' (str): prompt di input standardizzato.
                  * '_view._lunghezza_min' (int): lunghezza minima consentita per il titolo.
                  * '_view._lunghezza_max' (int): lunghezza massima consentita per il titolo.
                  * '_view._mostra_feedback_errore_lunghezza_titolo' (callable): messaggio errore lunghezza.
                  * '_view._mostra_feedback_errore_estremi_titolo' (callable): messaggio errore estremi.
                  * '_view._mostra_feedback_errore_cifre_titolo' (callable): messaggio errore cifre.
                  * '_view._mostra_feedback_errore_caratteri_titolo' (callable): messaggio errore caratteri.
                  * '_view._mostra_feedback_errore_titolo_duplicato' (callable): messaggio errore duplicato.
        """
        # str - caratteri di punteggiatura consentiti nel titolo.
        punteggiatura_consentita = ":'?"
        
        # set - insieme dei titoli esistenti per controllo duplicati (case-insensitive).
        titoli_esistenti = {
            articolo["Titolo"].strip().casefold()
            for articoli in self._view._dataset_articoli.values()
            for articolo in articoli
        }
        
        while True:
            # Acquisisce l'input dell'utente per il titolo.
            scelta = self._input(self._view._input_prompt).strip()
            
            # Verifica e gestisce comandi di navigazione speciali.
            if scelta in self._SCELTE_SPECIALI:
                return self._gestisci_scelta_speciale(scelta)
            else:
                titolo = scelta.strip()
                
                # Controllo lunghezza del titolo.
                if not (self._view._lunghezza_min <= len(titolo) <= self._view._lunghezza_max):
                    self._view._mostra_feedback_errore_lunghezza_titolo()
                    continue
                
                # Controllo caratteri agli estremi del titolo.
                if not titolo[0].isalpha() or not titolo[-1].isalpha():
                    self._view._mostra_feedback_errore_estremi_titolo()
                    continue
                
                # Controllo presenza di cifre numeriche.
                if any(char.isdigit() for char in titolo):
                    self._view._mostra_feedback_errore_cifre_titolo()
                    continue
                    
                # Controllo caratteri consentiti.
                if not all(char.isalpha() or char in punteggiatura_consentita or char == ' ' for char in titolo):
                    self._view._mostra_feedback_errore_caratteri_titolo()
                    continue
                    
                # Controllo titolo duplicato (case-insensitive).
                if titolo.casefold() in titoli_esistenti:
                    self._view._mostra_feedback_errore_titolo_duplicato()
                    continue
                
                # Normalizza il titolo con prima lettera maiuscola.
                titolo = titolo[0].upper() + titolo[1:] 
                return titolo
```

```python
class HomeController(DatasetArticoliController):
    def __init__(self, dataset_articoli: dict, funzione_input : callable = input):
        """
        Controller principale e punto di ingresso unico dell'applicazione.
        
        Coordina tutti i controller specializzati e funge da interfaccia
        centralizzata per l'utente, semplificando l'accesso a tutte le
        funzionalità del sistema.

        Args:
            - funzione_input: funzione per gestire l'input dell'utente (default: input built-in).
            
            - dataset_articoli: dizionario contenente gli articoli organizzati per categoria.
                Struttura attesa:
                {
                    "Categoria1": [
                        {
                            "Posizionamento": int,
                            "TipoNarrativa": str, 
                            "Titolo": str
                        },
                        ...
                    ],
                    "Categoria2": [...],
                    ...
                }
        """
        super().__init__(dataset_articoli, funzione_input)
        
        # ATTRIBUTI DI COMPOSIZIONE.
        # Composizione con i controller specializzati per ogni funzionalità.
        # Ogni controller gestisce un dominio specifico dell'applicazione.
        self._ricerca_unitaria_controller = RicercaUnitariaController(dataset_articoli, funzione_input)
        self._ricerca_combinata_controller = RicercaCombinataController(dataset_articoli, funzione_input)
        self._registrazione_controller = RegistrazioneController(dataset_articoli, funzione_input)
        self._visualizzazione_controller = VisualizzazioneController(dataset_articoli, funzione_input)
        self._analisi_controller = AnalisiController(dataset_articoli, funzione_input)
   
    def avvia(self) -> None | ExitProgram:
        """
        Punto di ingresso principale dell'applicazione.
        
        Gestisce il ciclo di vita completo del programma, coordinando la
        navigazione tra i vari menu e delegando le operazioni specifiche
        ai controller specializzati.

        Returns:
            - ExitProgram: quando l'utente richiede la terminazione dell'applicazione.
            - None: quando il ciclo principale viene interrotto normalmente.

        Note:
            - Utilizza:
              * '_scelta_uscita': codice speciale per la terminazione del programma.
              * '_view._mostra_menu_principale': metodo per visualizzare il menu iniziale.
              * '_view._menu_principale': configurazione del menu principale.
            
              * '_elabora_input_menu': metodo di gestione acquisizione e validazione input utente.

                - Controller compositi:
                * '_ricerca_unitaria_controller': gestione ricerche per criterio singolo.
                * '_ricerca_combinata_controller': gestione ricerche per criteri multipli.
                * '_registrazione_controller': gestione inserimento nuovi articoli.
                * '_visualizzazione_controller': gestione visualizzazioni complete.
                * ' _analisi_controller': gestione analisi statistiche.
        """
        # dict - Mappatura tra selezione utente e metodi controller corrispondenti.
        # Ogni chiave numerica corrisponde a una funzionalità specifica.
        azioni = {
            '1': self._ricerca_unitaria_controller._ricerca_criterio_singolo,
            '2': self._ricerca_combinata_controller._ricerca_criteri_multipli,
            '3': self._visualizzazione_controller._visualizza_dataset_completo,
            '4': self._registrazione_controller._inserisci_nuovo_articolo,
            '5': self._analisi_controller._visualizza_dati_statistici
        }
        
        # Ciclo principale dell'applicazione.
        while True:
            # Gestisce l'interazione con il menu principale e validazione input.
            scelta = self._elabora_input_menu(
                mostra_interfaccia=self._view._mostra_menu_principale,
                menu=self._view._menu_principale,
                scelte_speciali=[self._scelta_uscita]
            )

            # Gestisce richiesta di terminazione dell'applicazione.
            if isinstance(scelta, ExitProgram):
                return ExitProgram()
            
            # Esegue l'azione corrispondente alla selezione utente.
            # 'risultato' corrisponderà a None se il metodo ha terminato normalmente, con nessun codice di stato.
            risultato = azioni[scelta]()
            
            # Gestisce i comandi di navigazione ritornati dai controller figli.
            if isinstance(risultato, ExitProgram):
                break # Uscita dal programma.
            elif isinstance(risultato, (ReturnToNearestMenu, GoToMainMenu)):
                continue # Ristampa il menù superiore (il menù principale).
```

### Scenario Testing.

#### Introduzione.

In questa sezione è definita la funzione che costituisce il sistema di testing    
basato su scenari interattivi completi.  

La funzione `esegui_scenario()` crea un ambiente di verifica controllato e    
riproducibile che simula fedelmente l'interazione di un utente reale con    
l'applicazione. Questo approccio permette di testare l'intero flusso di   
navigazione, inclusa la gestione degli errori e l'uso dei comandi speciali   
(-1, 0, 00). 

#### 1. Metodo per test automatizzati.

```python
def esegui_scenario(dataset: dict, input_simulato: list[list[str]]) -> None:
    """
    Esegue uno scenario di test simulato interattivo completo.
    
    Crea un ambiente di test controllato che simula l'interazione utente con l'applicazione,
    utilizzando input predefiniti invece di input reale da tastiera. 
    In questo modo si possono testare automaticamente flussi complessi, sempre nelle stesse condizioni.
    
    Il metodo:
    1. Configura un controller con input mock,
    2. Simula sequenzialmente tutti gli input utente specificati,
    3. Registra tutto il flusso di interazione per analizzarlo successivamente.
    
    Args:
        - dataset (dict): Il dataset strutturato da passare al controller per l'esecuzione.
                          Deve rispettare la struttura standard del progetto.
        - input_simulato (list[list[str]]): Sequenza di input simulati da fornire al controller.
                                            Ogni elemento è una lista [valore_input, descrizione] dove:
                                            - valore_input: stringa da passare come input,
                                            - descrizione: testo descrittivo dell'azione simulata.
                                         
    Note:
        - Utilizza la classe 'HomeController' come punto di ingresso dell'applicazione.
        - Implementa un generatore di input mock che fornisce valori in sequenza prefissata.
        - Visualizza chiaramente sia gli input simulati che le risposte del sistema.
        - Termina automaticamente quando tutti gli input sono stati consumati.
        - Mostra separatori visivi per distinguere chiaramente gli scenari durante l'esecuzione.
        
    Utilizzo tipico:
    1. Definire la sequenza di input simulati con relative descrizioni.
    2. Richiamare 'esegui_scenario()' con dataset e input predefiniti.
    3. Verificare che l'output corrisponda alle aspettative.

    Esempio pratico:
    
    esegui_scenario(
        dataset=dataset_di_test,
        input_simulato=[
            ['1', "1. Seleziona operazione di ricerca"],
            ['1', "2. Seleziona criterio 'per categoria'"],
            ['12', "3. Test: categoria fuori range (dovrebbe fallire)"],
            ['2', "4. Test: categoria valida (dovrebbe funzionare)"],
            ['-1', "5. Test: navigazione 'torna indietro'"],
            ['00', "6. Test: uscita dal programma"]
        ]
    )    
    """
    LINEA_SEPARATRICE = "[cyan]" + '-' * 120 + "[/cyan]\n"
    
    console.print(LINEA_SEPARATRICE)
    
    scenario = iter(input_simulato)

    def input_mock(prompt=''):
        """
        Funzione mock che sostituisce l'input standard durante i test.
        
        Fornisce input predefiniti in sequenza invece di leggere da tastiera,
        permettendo di automatizzare scenari di interazione complessi.
        
        Args:
            prompt (str): Prompt originale del sistema (non utilizzato nella simulazione).
            
        Returns:
            str: Il prossimo valore di input dalla sequenza simulata.
            
        Note:
            - Stampa la descrizione dell'azione simulata per tracciabilità.
            - Mostra il valore di input effettivo fornito al sistema.
            - Si ferma automaticamente quando la sequenza di input è esaurita.
        """
        console.print('\n')
        valore = next(scenario)  # Prende il prossimo input dalla sequenza.
        if len(valore) == 2:
            console.print(f"[bold]{valore[1]}[/bold]") # Mostra descrizione azione.
        console.print(f"[italic]Input inserito:[/italic] {valore[0]}")
        return valore[0]
    
    controller = HomeController(dataset_articoli=dataset, funzione_input=input_mock)
    controller.avvia()
    
    console.print(LINEA_SEPARATRICE)
```
